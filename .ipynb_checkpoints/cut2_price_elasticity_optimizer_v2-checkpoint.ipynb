## STEP 1: Business Understanding

**ปัญหา:** ร้าน Retail SME จัดโปรโมชั่นแบบเดาสุ่ม
- ลด 20% แต่ยอดขายไม่ขึ้น → Margin หายเปล่า
- ลดแค่ 5% ก็ขายดีขึ้นมาก → แต่กลับไม่จัดโปร เสียโอกาส

**Solution:** ใช้ Price Elasticity บอกว่า "ลดเท่าไหร่ถึงคุ้ม"

| KPI | นิยาม | เป้าหมาย |
|-----|--------|----------|
| Gross Profit Margin | (Revenue − COGS) / Revenue | เพิ่มขึ้น vs baseline |
| Promotion ROI | Revenue Uplift / Discount Cost | > 1.0 |
| Revenue Uplift | % เปลี่ยนแปลงของยอดขาย vs ก่อนโปร | บวกและมีนัยสำคัญ |

## STEP 3: Data Cleaning & Preparation

```python
# ── 3.1 Clean sales ──────────────────────────────────────────
print(f'Before: {sales.shape}')

sales = sales.drop_duplicates().reset_index(drop=True)
sales['datetime'] = pd.to_datetime(sales['datetime'], errors='coerce')
sales = sales.dropna(subset=['price'])
sales = sales[sales['qty'] > 0]                        # ตัด returns
sales = sales[~sales['qty'].isin([500, 888, 999])]     # ตัด outlier qty
sales['revenue']       = sales['price'] * sales['qty']
sales['has_promotion'] = sales['promotion_id'].notna().astype(int)

print(f'After : {sales.shape}')
```

```python
# ── 3.2 Clean product ────────────────────────────────────────
print(f'Before: {product.shape}')

product = product.drop_duplicates().reset_index(drop=True)
product['product_taxonomies'] = (
    product['product_taxonomies'].fillna('unknown').str.strip().str.lower()
)
product['price'] = (
    product['price'].astype(str)
    .str.replace(r'[^\d.]', '', regex=True)
)
product['price'] = pd.to_numeric(product['price'], errors='coerce')

print(f'After : {product.shape}')
```

```python
# ── 3.3 Clean promo ──────────────────────────────────────────
promo['start_date'] = pd.to_datetime(promo['start_date'], errors='coerce')
promo['end_date']   = pd.to_datetime(promo['end_date'],   errors='coerce')

# buy1get1 = ลูกค้าจ่ายแค่ครึ่งราคา = discount 50%
promo.loc[promo['promotion_type'] == 'buy1get1', 'discount'] = 0.50

print('Promo types & discount:')
print(promo.groupby('promotion_type')['discount'].describe().round(2))
```

```python
# ── 3.4 Merge ────────────────────────────────────────────────
df = sales.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id', how='left', suffixes=('_sold', '_list')
)
df = df.merge(
    promo[['promotion_id', 'discount', 'promotion_type']],
    on='promotion_id', how='left'
)

# Feature Engineering
df['discount']        = df['discount'].fillna(0.0)
df['effective_price'] = df['price_list'] * (1 - df['discount'])
df['gross_profit']    = (df['price_sold'] - df['cogs']) * df['qty']

print(f'Master table: {df.shape}')
df[['product_id','price_sold','discount','effective_price','qty','revenue','gross_profit']].head()
```

```python
# ── EDA Post-Merge: Revenue by Category ─────────────────────
rev_by_cat = df.groupby('product_taxonomies')['revenue'].sum().sort_values()

fig, ax = plt.subplots(figsize=(9, 5))
rev_by_cat.plot(kind='barh', ax=ax, color='steelblue')
ax.set_title('Total Revenue by Product Category')
ax.set_xlabel('Revenue (Baht)')
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.tight_layout()
plt.show()
```

```python
# ── EDA: Discount vs Avg Qty ─────────────────────────────────
df['discount_bin'] = pd.cut(
    df['discount'],
    bins=[-0.01, 0, 0.10, 0.20, 0.30, 0.50, 1.0],
    labels=['0%', '1-10%', '11-20%', '21-30%', '31-50%', '>50%']
)

avg_qty = df.groupby('discount_bin', observed=True)['qty'].mean().reset_index()

fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(avg_qty['discount_bin'], avg_qty['qty'], color='coral', edgecolor='white')
ax.set_title('Average Qty Sold by Discount Level')
ax.set_xlabel('Discount Range')
ax.set_ylabel('Avg Qty')
plt.tight_layout()
plt.show()

print(avg_qty)
```

## STEP 4: Modeling — Price Elasticity

**แนวคิด: Price Elasticity of Demand (PED)**

$$PED = \frac{\% \Delta Quantity}{\% \Delta Price}$$

- PED < -1 → **Elastic** (ลดราคาแล้วยอดขายขึ้นมาก → ควรจัดโปร)
- -1 < PED < 0 → **Inelastic** (ลดราคาแต่ยอดขายขึ้นน้อย → ไม่ควรลดมาก)

**วิธีประมาณ:** Log-Log OLS Regression ระดับสินค้า
$$\ln(Qty) = \alpha + \beta_1 \cdot \ln(Price) + \varepsilon$$

- β₁ คือ elasticity โดยตรง
- สินค้าที่มีข้อมูลน้อย (<5 transactions) → ใช้ค่า **category median** แทน
- Shrinkage: `elasticity = 0.7 × category_mean + 0.3 × product_estimate` เพื่อลด noise

```python
# ── 4.1 Compute Elasticity per Product ───────────────────────
def compute_elasticity(group):
    """Log-Log OLS: ln(qty) ~ ln(price_sold). คืน slope = elasticity"""
    log_p = np.log(group['price_sold'])
    log_q = np.log(group['qty'])
    mask  = np.isfinite(log_p) & np.isfinite(log_q)
    if mask.sum() < 5:
        return np.nan
    slope, *_ = stats.linregress(log_p[mask], log_q[mask])
    return slope

elast_df = (
    df.groupby('product_id')
    .apply(compute_elasticity)
    .reset_index()
    .rename(columns={0: 'elasticity_raw'})
)
elast_df = elast_df.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id', how='left'
)

# Shrinkage: ผสม product estimate กับ category mean เพื่อลด noise
cat_mean = elast_df.groupby('product_taxonomies')['elasticity_raw'].mean()
elast_df['elasticity'] = elast_df.apply(
    lambda r: (
        cat_mean.get(r['product_taxonomies'], -1.0)          # ข้อมูลน้อยใช้ category
        if pd.isna(r['elasticity_raw'])
        else 0.7 * cat_mean.get(r['product_taxonomies'], -1.0)
             + 0.3 * r['elasticity_raw']                     # shrinkage
    ),
    axis=1
)

elast_df['elasticity_type'] = elast_df['elasticity'].apply(
    lambda e: 'Elastic' if e < -1 else ('Inelastic' if e < 0 else 'Insensitive')
)

print('Elasticity Type Distribution:')
print(elast_df['elasticity_type'].value_counts())
print('\nElasticity Stats:')
print(elast_df['elasticity'].describe().round(3))
```

```python
# ── 4.2 Visualize Elasticity ─────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].hist(elast_df['elasticity'].dropna(), bins=25, color='steelblue', edgecolor='white')
axes[0].axvline(-1, color='red',    linestyle='--', label='= -1 (unit elastic)')
axes[0].axvline(0,  color='orange', linestyle='--', label='= 0')
axes[0].set_title('Price Elasticity Distribution')
axes[0].set_xlabel('Elasticity Coefficient')
axes[0].set_ylabel('# Products')
axes[0].legend(fontsize=8)

cat_e  = elast_df.groupby('product_taxonomies')['elasticity'].mean().sort_values()
colors = ['#388e3c' if e < -1 else '#d32f2f' for e in cat_e]
axes[1].barh(cat_e.index, cat_e.values, color=colors)
axes[1].axvline(-1, color='black', linestyle='--', linewidth=0.8)
axes[1].set_title('Avg Elasticity by Category\n(Green = Elastic = ควรจัดโปร)')
axes[1].set_xlabel('Avg Elasticity')

plt.tight_layout()
plt.show()
```

```python
# ── 4.3 Optimal Discount Finder ──────────────────────────────
def find_optimal_discount(base_price, cogs, elasticity, optimize='gross_profit'):
    """
    Demand model: Q(d) = Q0 × (1-d)^elasticity  (Power Law)
    Revenue(d)     = price×(1-d) × Q(d)
    GrossProfit(d) = (price×(1-d) - cogs) × Q(d)

    ลอง discount 8 ระดับ: [0,5,10,15,20,30,40,50]%
    """
    discounts = [0, 0.05, 0.10, 0.15, 0.20, 0.30, 0.40, 0.50]
    base_qty  = 100

    rows = []
    for d in discounts:
        price_d = base_price * (1 - d)
        qty_d   = base_qty * ((1 - d) ** elasticity)
        rows.append({
            'discount'    : d,
            'revenue'     : price_d * qty_d,
            'gross_profit': (price_d - cogs) * qty_d
        })

    res      = pd.DataFrame(rows)
    best_idx = res[optimize].idxmax()
    return res, best_idx

print('find_optimal_discount() ready')
```

```python
# ── 4.4 Visualize Discount Curve (2 ตัวอย่าง) ────────────────
elastic_p   = elast_df[elast_df['elasticity'] < -1].dropna(subset=['price','cogs']).iloc[0]
inelastic_p = elast_df[elast_df['elasticity'].between(-1, 0)].dropna(subset=['price','cogs']).iloc[0]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, prod, label in [
    (axes[0], elastic_p,   'Elastic Product'),
    (axes[1], inelastic_p, 'Inelastic Product')
]:
    res, best_idx = find_optimal_discount(prod['price'], prod['cogs'], prod['elasticity'])
    ax.plot(res['discount']*100, res['gross_profit'],
            'o-', color='steelblue', label='Gross Profit')
    ax.plot(res['discount']*100, res['revenue'],
            's--', color='coral', label='Revenue')
    ax.axvline(res.loc[best_idx, 'discount']*100, color='green',
               linestyle=':', linewidth=2,
               label=f"Optimal = {res.loc[best_idx,'discount']:.0%}")
    ax.set_title(f'{label} | {prod["product_id"]}\nElasticity = {prod["elasticity"]:.2f} ({prod["elasticity_type"]})')
    ax.set_xlabel('Discount (%)')
    ax.set_ylabel('Value (Index, baseline=100 units)')
    ax.legend(fontsize=8)

plt.suptitle('Gross Profit & Revenue vs Discount Level', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ── 4.5 Generate Recommendations for All Products ────────────
cur_disc = (
    df[df['discount'] > 0]
    .groupby('product_id')['discount'].mean()
    .reset_index()
    .rename(columns={'discount': 'current_discount'})
)

recs = []
for _, row in elast_df.iterrows():
    pid, price, cogs, elast = row['product_id'], row['price'], row['cogs'], row['elasticity']
    if pd.isna(price) or pd.isna(cogs) or pd.isna(elast):
        continue

    res, best_idx = find_optimal_discount(price, cogs, elast, optimize='gross_profit')
    best = res.loc[best_idx]

    cur_d_row = cur_disc[cur_disc['product_id'] == pid]
    cur_d     = cur_d_row['current_discount'].values[0] if len(cur_d_row) > 0 else 0.0

    cur_idx    = res['discount'].sub(cur_d).abs().idxmin()
    cur        = res.loc[cur_idx]
    gp_change  = (best['gross_profit'] - cur['gross_profit']) / (abs(cur['gross_profit']) + 1e-9)
    rev_change = (best['revenue']      - cur['revenue'])      / (cur['revenue']           + 1e-9)

    # Promo ROI: Revenue Uplift / Discount Cost
    rev_uplift    = best['revenue'] - cur['revenue']
    discount_cost = best['revenue'] * best['discount']
    promo_roi     = rev_uplift / (discount_cost + 1e-9)

    if   best['discount'] > cur_d + 0.03: action = 'Increase Discount'
    elif best['discount'] < cur_d - 0.03: action = 'Reduce Discount'
    else:                                  action = 'Keep / Stop Promotion'

    recs.append({
        'product_id'           : pid,
        'category'             : row['product_taxonomies'],
        'elasticity'           : round(elast, 3),
        'elasticity_type'      : row['elasticity_type'],
        'list_price'           : round(price, 2),
        'cogs'                 : round(cogs, 2),
        'current_discount'     : round(cur_d, 2),
        'recommended_discount' : round(best['discount'], 2),
        'expected_gp_change'   : round(gp_change, 4),
        'expected_rev_change'  : round(rev_change, 4),
        'promo_roi'            : round(promo_roi, 2),
        'action'               : action
    })

reco_df = pd.DataFrame(recs)
print(f'Recommendations generated: {len(reco_df)} products')
print('\nAction Summary:')
print(reco_df['action'].value_counts())
```

## STEP 5: Evaluation

ประเมิน 3 วิธี:
1. **A/B Comparison** — เปรียบ avg qty/revenue ช่วงมีโปร vs ไม่มีโปร
2. **Promotion ROI** — revenue uplift / discount cost (> 1.0 = คุ้ม)
3. **Baseline Comparison** — optimized vs current practice

```python
# ── 5.1 A/B Comparison: Promo vs Non-Promo ───────────────────
promo_products = df[df['has_promotion'] == 1]['product_id'].unique()

ab_rows = []
for pid in promo_products:
    sub  = df[df['product_id'] == pid]
    with_ = sub[sub['has_promotion'] == 1]
    without = sub[sub['has_promotion'] == 0]
    if len(with_) < 3 or len(without) < 3:
        continue

    qty_uplift = (with_['qty'].mean() - without['qty'].mean()) / (without['qty'].mean() + 1e-9)
    rev_uplift = (with_['revenue'].mean() - without['revenue'].mean()) / (without['revenue'].mean() + 1e-9)

    ab_rows.append({
        'product_id'      : pid,
        'avg_discount'    : round(with_['discount'].mean(), 2),
        'qty_no_promo'    : round(without['qty'].mean(), 1),
        'qty_with_promo'  : round(with_['qty'].mean(), 1),
        'qty_uplift'      : f"{qty_uplift:+.1%}",
        'rev_uplift'      : f"{rev_uplift:+.1%}",
        'promo_beneficial': 'YES' if rev_uplift > 0 else 'NO'
    })

ab_df = pd.DataFrame(ab_rows)
beneficial_pct = (ab_df['promo_beneficial'] == 'YES').mean()
print(f'Products validated: {len(ab_df)}')
print(f'Promotion was revenue-beneficial in {beneficial_pct:.0%} of tested products')
print('\nSample:')
ab_df.head(10)
```

```python
# ── 5.2 Promotion ROI per Product ────────────────────────────
roi_df = reco_df[['product_id', 'recommended_discount', 'promo_roi']].copy()

fig, ax = plt.subplots(figsize=(8, 5))
roi_sorted = roi_df.sort_values('promo_roi')
colors = ['#d32f2f' if r < 1 else '#388e3c' for r in roi_sorted['promo_roi']]
ax.barh(roi_sorted['product_id'], roi_sorted['promo_roi'].clip(-3, 5), color=colors)
ax.axvline(1, color='black', linestyle='--', label='Break-even ROI = 1.0')
ax.set_title('Promotion ROI by Product\n(Green = ROI > 1 = คุ้ม | Red = ไม่คุ้มลด)')
ax.set_xlabel('Promotion ROI')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

print(f"Avg Promo ROI       : {roi_df['promo_roi'].mean():.2f}")
print(f"Products ROI > 1.0  : {(roi_df['promo_roi'] > 1).sum()} / {len(roi_df)}")
```

```python
# ── 5.3 Baseline Comparison: Current vs Optimized ────────────
reduce   = reco_df[reco_df['action'] == 'Reduce Discount']
increase = reco_df[reco_df['action'] == 'Increase Discount']
keep     = reco_df[reco_df['action'] == 'Keep / Stop Promotion']

print('=== Recommendation Summary ===')
print(f'  Reduce Discount   : {len(reduce):>3} products')
print(f'  Increase Discount : {len(increase):>3} products')
print(f'  Keep / Stop       : {len(keep):>3} products')

avg_gp_uplift = reco_df['expected_gp_change'].mean()
print(f'\nAvg Expected GP Change (vs current): {avg_gp_uplift:+.1%}')

print('\n=== Top 10 ควร Reduce Discount (ลดโปรได้ ไม่เสีย Margin) ===')
print(
    reduce[['product_id','category','elasticity_type',
            'current_discount','recommended_discount','expected_gp_change']]
    .assign(current_discount=lambda x: x['current_discount'].apply(lambda v: f"{v:.0%}"),
            recommended_discount=lambda x: x['recommended_discount'].apply(lambda v: f"{v:.0%}"),
            expected_gp_change=lambda x: x['expected_gp_change'].apply(lambda v: f"{v:+.1%}"))
    .sort_values('current_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)

print('\n=== Top 10 ควร Increase Discount (จัดโปรเพิ่ม → กำไรขึ้น) ===')
print(
    increase[['product_id','category','elasticity_type',
              'current_discount','recommended_discount','expected_rev_change']]
    .assign(current_discount=lambda x: x['current_discount'].apply(lambda v: f"{v:.0%}"),
            recommended_discount=lambda x: x['recommended_discount'].apply(lambda v: f"{v:.0%}"),
            expected_rev_change=lambda x: x['expected_rev_change'].apply(lambda v: f"{v:+.1%}"))
    .sort_values('recommended_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)
```

## สรุป (Limitations & Assumptions)

**Assumptions:**
1. Log-Log relationship — สมมติว่า Demand function เป็น Power Law
2. buy1get1 = discount จริง 50%
3. Shrinkage weight 70/30 (category/product) — ช่วยลด noise สินค้าที่ข้อมูลน้อย

**Limitations:**
- Elasticity เป็น correlation ไม่ใช่ true causal effect
- ข้อมูล 1 ปี → อาจยังไม่เสถียร
- ไม่ได้คำนึง seasonality, cross-product effects, stock constraint
- base_qty = 100 เป็น index ไม่ใช่ตัวเลขจริง → ใช้เปรียบเทียบ % เท่านั้น

**KPI ที่วัดได้:**

| KPI | วัดจาก |
|-----|--------|
| Gross Profit Change | `expected_gp_change` ใน recommendation table |
| Promotion ROI | `promo_roi` (> 1.0 = คุ้ม) |
| Revenue Uplift | `expected_rev_change` |

**Next Steps:**
- เพิ่ม seasonality effect ด้วย month/week feature
- A/B test จริงกับ 5–10 สินค้า pilot ก่อน
- อาจเพิ่ม store-level optimization ในอนาคต
