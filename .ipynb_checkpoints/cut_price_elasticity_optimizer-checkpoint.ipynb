# 💡 Price Elasticity Promotion Optimizer
**เป้าหมาย:** ลดราคาเท่าไหร่ → ทำให้กำไร/รายได้สูงสุด

**Flow (3 ขั้นตอนหลัก):**
1. โหลด + ทำความสะอาดข้อมูล
2. คำนวณ Price Elasticity ต่อสินค้า
3. แนะนำส่วนลดที่เหมาะสม + แสดงผล

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
```

---
## STEP 1: โหลด & ทำความสะอาดข้อมูล

```python
# โหลดข้อมูล
sales   = pd.read_csv('./data/sales_transaction.csv')
product = pd.read_csv('./data/product_master.csv')
promo   = pd.read_csv('./data/promotion_master.csv')

print('Raw shapes:')
print(f'  sales   : {sales.shape}')
print(f'  product : {product.shape}')
print(f'  promo   : {promo.shape}')
```

```python
# ── Clean sales ──────────────────────────────────────────────
sales = sales.drop_duplicates().reset_index(drop=True)
sales['datetime'] = pd.to_datetime(sales['datetime'], errors='coerce')
sales = sales.dropna(subset=['price'])
sales = sales[sales['qty'] > 0]                          # ตัด returns
sales = sales[~sales['qty'].isin([500, 888, 999])]       # ตัด outlier qty
sales['revenue'] = sales['price'] * sales['qty']

# ── Clean product ─────────────────────────────────────────────
product = product.drop_duplicates().reset_index(drop=True)
product['product_taxonomies'] = (
    product['product_taxonomies']
    .fillna('unknown')
    .str.strip().str.lower()
)
product['price'] = (
    product['price'].astype(str)
    .str.replace(r'[^\d.]', '', regex=True)
)
product['price'] = pd.to_numeric(product['price'], errors='coerce')

# ── Clean promo ───────────────────────────────────────────────
promo['start_date'] = pd.to_datetime(promo['start_date'], errors='coerce')
promo['end_date']   = pd.to_datetime(promo['end_date'],   errors='coerce')
# buy1get1 = discount จริง 50%
promo.loc[promo['promotion_type'] == 'buy1get1', 'discount'] = 0.50

print('Cleaned shapes:')
print(f'  sales   : {sales.shape}')
print(f'  product : {product.shape}')
print(f'  promo   : {promo.shape}')
```

```python
# ── Merge ─────────────────────────────────────────────────────
df = sales.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id', how='left', suffixes=('_sold', '_list')
)
df = df.merge(
    promo[['promotion_id', 'discount', 'promotion_type']],
    on='promotion_id', how='left'
)

# fill discount = 0 ถ้าไม่มีโปร
df['discount'] = df['discount'].fillna(0.0)

# effective price = ราคาที่ลูกค้าจ่ายจริง
df['effective_price'] = df['price_list'] * (1 - df['discount'])

# gross profit ต่อ transaction
df['gross_profit'] = (df['price_sold'] - df['cogs']) * df['qty']

print(f'Master table shape: {df.shape}')
df[['product_id','price_sold','discount','effective_price','qty','revenue','gross_profit']].head()
```

---
## STEP 2: คำนวณ Price Elasticity

ใช้ **Log-Log OLS Regression** ระดับสินค้า:

$$\ln(Qty) = \alpha + \beta_1 \cdot \ln(Price) + \varepsilon$$

- β₁ = Elasticity
- ถ้า β₁ < -1 → **Elastic** (ลดราคาแล้วยอดขายขึ้นมาก → ควรจัดโปร)
- ถ้า -1 < β₁ < 0 → **Inelastic** (ลดราคาแต่ยอดขายขึ้นน้อย → ไม่ควรลดมาก)

```python
def compute_elasticity(group):
    """Log-Log OLS: ln(qty) ~ ln(price_sold). คืน slope = elasticity"""
    log_p = np.log(group['price_sold'])
    log_q = np.log(group['qty'])
    mask  = np.isfinite(log_p) & np.isfinite(log_q)
    if mask.sum() < 5:
        return np.nan
    slope, *_ = stats.linregress(log_p[mask], log_q[mask])
    return slope

# คำนวณ elasticity ทุก product
elast_df = (
    df.groupby('product_id')
    .apply(compute_elasticity)
    .reset_index()
    .rename(columns={0: 'elasticity'})
)

# join กับ product info
elast_df = elast_df.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id', how='left'
)

# สินค้าที่ข้อมูลน้อย → ใช้ median ของ category แทน
cat_median = elast_df.groupby('product_taxonomies')['elasticity'].median()
elast_df['elasticity'] = elast_df.apply(
    lambda r: cat_median.get(r['product_taxonomies'], -1.0)
    if pd.isna(r['elasticity']) else r['elasticity'],
    axis=1
)

elast_df['elasticity_type'] = elast_df['elasticity'].apply(
    lambda e: 'Elastic' if e < -1 else ('Inelastic' if e < 0 else 'Insensitive')
)

print(f'Products with elasticity: {elast_df["elasticity"].notna().sum()}')
print(elast_df['elasticity_type'].value_counts())
print('\nElasticity stats:')
print(elast_df['elasticity'].describe().round(3))
```

```python
# ── Plot: Elasticity Distribution + by Category ───────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].hist(elast_df['elasticity'].dropna(), bins=25,
             color='steelblue', edgecolor='white')
axes[0].axvline(-1, color='red',    linestyle='--', label='Elasticity = -1 (unit elastic)')
axes[0].axvline(0,  color='orange', linestyle='--', label='Elasticity = 0')
axes[0].set_title('Price Elasticity Distribution')
axes[0].set_xlabel('Elasticity Coefficient')
axes[0].set_ylabel('# Products')
axes[0].legend(fontsize=8)

cat_e  = elast_df.groupby('product_taxonomies')['elasticity'].mean().sort_values()
colors = ['#388e3c' if e < -1 else '#d32f2f' for e in cat_e]
axes[1].barh(cat_e.index, cat_e.values, color=colors)
axes[1].axvline(-1, color='black', linestyle='--', linewidth=0.8)
axes[1].set_title('Avg Elasticity by Category\n(Green = Elastic = ควรจัดโปร)')
axes[1].set_xlabel('Avg Elasticity')

plt.tight_layout()
plt.show()
```

---
## STEP 3: หา Optimal Discount & สร้าง Recommendation Table

```python
def optimal_discount(base_price, cogs, elasticity, optimize='gross_profit'):
    """
    คำนวณ Revenue & Gross Profit ที่ discount ต่างๆ (0-50%)
    แล้วคืน discount ที่ให้ผลลัพธ์ดีที่สุด
    
    qty(d) = qty_base × (1-d)^elasticity  (power-law demand)
    """
    discounts = np.arange(0, 0.51, 0.01)
    base_qty  = 100

    rows = []
    for d in discounts:
        price_d = base_price * (1 - d)
        qty_d   = base_qty * ((1 - d) ** elasticity)
        rows.append({
            'discount'     : d,
            'revenue'      : price_d * qty_d,
            'gross_profit' : (price_d - cogs) * qty_d
        })

    res = pd.DataFrame(rows)
    best_idx = res[optimize].idxmax()
    return res, best_idx


# discount ปัจจุบันของแต่ละสินค้า (average จากที่เคยมีโปร)
cur_disc = (
    df[df['discount'] > 0]
    .groupby('product_id')['discount'].mean()
    .reset_index()
    .rename(columns={'discount': 'current_discount'})
)

# สร้าง recommendation ทุก product
recs = []
for _, row in elast_df.iterrows():
    pid, price, cogs, elast = row['product_id'], row['price'], row['cogs'], row['elasticity']
    if pd.isna(price) or pd.isna(cogs) or pd.isna(elast):
        continue

    res, best_idx = optimal_discount(price, cogs, elast, optimize='gross_profit')
    best = res.loc[best_idx]

    cur_d_row = cur_disc[cur_disc['product_id'] == pid]
    cur_d     = cur_d_row['current_discount'].values[0] if len(cur_d_row) > 0 else 0.0

    # เทียบ best vs current
    cur_idx = res['discount'].sub(cur_d).abs().idxmin()
    cur     = res.loc[cur_idx]

    gp_change  = (best['gross_profit'] - cur['gross_profit']) / (abs(cur['gross_profit']) + 1e-9)
    rev_change = (best['revenue']      - cur['revenue'])      / (cur['revenue']           + 1e-9)

    if   best['discount'] > cur_d + 0.03: action = 'Increase Discount'
    elif best['discount'] < cur_d - 0.03: action = 'Reduce Discount'
    else:                                  action = 'Keep / Stop Promotion'

    recs.append({
        'product_id'           : pid,
        'category'             : row['product_taxonomies'],
        'elasticity'           : round(elast, 3),
        'elasticity_type'      : row['elasticity_type'],
        'list_price'           : round(price, 2),
        'cogs'                 : round(cogs, 2),
        'current_discount'     : round(cur_d, 2),
        'recommended_discount' : round(best['discount'], 2),
        'expected_gp_change'   : f"{gp_change:+.1%}",
        'expected_rev_change'  : f"{rev_change:+.1%}",
        'action'               : action
    })

reco_df = pd.DataFrame(recs)
print(f'Recommendations for {len(reco_df)} products\n')
print(reco_df['action'].value_counts())
```

```python
# ── แสดงตาราง Recommendation ────────────────────────────────
display_cols = [
    'product_id', 'category', 'elasticity_type',
    'current_discount', 'recommended_discount',
    'expected_gp_change', 'expected_rev_change', 'action'
]

# format % สวยงาม
out = reco_df[display_cols].copy()
out['current_discount']     = out['current_discount'].apply(lambda x: f"{x:.0%}")
out['recommended_discount'] = out['recommended_discount'].apply(lambda x: f"{x:.0%}")

print('=== PROMOTION RECOMMENDATION TABLE ===')
out
```

```python
# ── Plot: Optimal Discount Curve (2 ตัวอย่าง) ────────────────
elastic_p   = reco_df[reco_df['elasticity'] < -1].iloc[0]
inelastic_p = reco_df[reco_df['elasticity'].between(-1, 0)].iloc[0]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, prod, label in [
    (axes[0], elastic_p,   'Elastic Product'),
    (axes[1], inelastic_p, 'Inelastic Product')
]:
    res, best_idx = optimal_discount(
        prod['list_price'], prod['cogs'], prod['elasticity']
    )
    ax.plot(res['discount']*100, res['gross_profit'],
            color='steelblue', label='Gross Profit')
    ax.plot(res['discount']*100, res['revenue'],
            color='coral', linestyle='--', label='Revenue')
    ax.axvline(res.loc[best_idx, 'discount']*100,
               color='green', linestyle=':', linewidth=2,
               label=f"Optimal = {res.loc[best_idx,'discount']:.0%}")
    ax.set_title(f'{label}  |  {prod["product_id"]}\nElasticity = {prod["elasticity"]:.2f}  ({prod["elasticity_type"]})')
    ax.set_xlabel('Discount (%)')
    ax.set_ylabel('Value (baseline 100 units)')
    ax.legend(fontsize=8)

plt.suptitle('Gross Profit & Revenue vs Discount Level', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ── Summary Dashboard (2×2) ───────────────────────────────────
fig, axes = plt.subplots(2, 2, figsize=(14, 9))
fig.suptitle('Price Elasticity Promotion Optimizer — Summary', fontsize=14, fontweight='bold')

COLORS = {'Reduce Discount': '#F44336', 'Increase Discount': '#4CAF50',
          'Keep / Stop Promotion': '#9E9E9E'}

# 1. Action Pie
act_cnt = reco_df['action'].value_counts()
axes[0,0].pie(act_cnt.values, labels=act_cnt.index, autopct='%1.1f%%',
              colors=[COLORS.get(a, 'grey') for a in act_cnt.index], startangle=90)
axes[0,0].set_title('Recommended Actions')

# 2. Avg Elasticity by Category
cat_e = reco_df.groupby('category')['elasticity'].mean().sort_values()
bar_c = ['#388e3c' if e < -1 else '#d32f2f' for e in cat_e]
axes[0,1].barh(cat_e.index, cat_e.values, color=bar_c)
axes[0,1].axvline(-1, color='black', linestyle='--', linewidth=0.8)
axes[0,1].set_title('Avg Elasticity by Category')
axes[0,1].set_xlabel('Elasticity')

# 3. Current vs Recommended Discount scatter
sc_colors = [COLORS.get(a, 'grey') for a in reco_df['action']]
axes[1,0].scatter(reco_df['current_discount']*100,
                  reco_df['recommended_discount']*100,
                  c=sc_colors, alpha=0.7, edgecolors='white')
axes[1,0].plot([0,50],[0,50],'k--', linewidth=0.8, label='No change')
axes[1,0].set_title('Current vs Recommended Discount')
axes[1,0].set_xlabel('Current Discount (%)')
axes[1,0].set_ylabel('Recommended Discount (%)')
axes[1,0].legend(fontsize=8)

# 4. Top 10 สินค้าที่ควร Increase Discount (missed opportunity)
top_inc = reco_df[reco_df['action'] == 'Increase Discount'] \
              .sort_values('recommended_discount', ascending=False).head(10)
axes[1,1].barh(top_inc['product_id'], top_inc['recommended_discount']*100,
               color='#4CAF50', edgecolor='white')
axes[1,1].set_title('Top Products: Increase Discount\n(Missed Opportunity)')
axes[1,1].set_xlabel('Recommended Discount (%)')

plt.tight_layout()
plt.savefig('summary_dashboard.png', dpi=120, bbox_inches='tight')
plt.show()
print('Saved: summary_dashboard.png')
```

```python
# ── Export CSV ────────────────────────────────────────────────
out.to_csv('promotion_recommendations.csv', index=False)
print('Saved: promotion_recommendations.csv')

print('\n=== Top 10 ควร Reduce Discount (ลดโปรได้ ไม่เสียอะไร) ===')
print(
    reco_df[reco_df['action'] == 'Reduce Discount']
    [['product_id','category','elasticity','current_discount','recommended_discount','expected_gp_change']]
    .sort_values('current_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)

print('\n=== Top 10 ควร Increase Discount (จัดโปรเพิ่ม → กำไรขึ้น) ===')
print(
    reco_df[reco_df['action'] == 'Increase Discount']
    [['product_id','category','elasticity','current_discount','recommended_discount','expected_gp_change']]
    .sort_values('recommended_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)
```

---
## สรุป

| Step | ทำอะไร | Output |
|------|---------|--------|
| 1 | โหลด + ทำความสะอาด | `df` master table |
| 2 | คำนวณ Price Elasticity | `elast_df` |
| 3 | หา Optimal Discount | `reco_df` + CSV + Dashboard |

**Assumptions:**
- Log-Log demand (Power Law) — เหมาะกับ Retail ทั่วไป
- สินค้าที่ข้อมูลน้อย (<5 transactions) ใช้ค่า median ของ category แทน
- buy1get1 = discount จริง 50%

**Limitations:**
- ข้อมูล 1 ปี → Elasticity อาจยังไม่เสถียร
- ไม่ได้คำนึง cross-product effects
- ยังไม่ได้แยก optimize ระดับ store
