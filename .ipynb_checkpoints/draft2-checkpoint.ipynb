## STEP 1 — Business Understanding
---


### 1.1 Business Problem
| | |
|---|---|
| **ประเภทธุรกิจ** | SME Retail ไทย (ร้านค้าปลีกขนาดเล็ก-กลาง) |
| **Core Problem** | จัด Promotion แบบ "หว่าน" ทำให้ Margin หายโดยไม่เพิ่ม Revenue จริง |
| **Root Cause** | ไม่รู้ว่าสินค้าไหน "sensitive" ต่อราคา — ลดแล้วขายดีขึ้น vs ลดแล้วแค่ขาดทุน |

### 1.2 KPI ที่ใช้วัดผล
| KPI | นิยาม | ทำไมเลือก |
|---|---|---|
| **Profit Margin %** | (Revenue - COGS) / Revenue × 100 | วัดว่าหลังจัดโปรยังมีกำไรมั้ย |
| **Promotion ROI** | Incremental Revenue / Discount Cost | วัดว่าโปรคุ้มค่ามั้ย |
| **Promotion Uplift %** | (Revenue with promo - Baseline) / Baseline | วัดว่าโปรช่วยเพิ่ม revenue จริงหรือแค่ cannibalize |

### 1.3 Expected Output
**Promotion Recommendation Table** — บอกเจ้าของร้านว่าสินค้าไหนควร:
- ลดราคาเพิ่ม (เพราะ demand sensitive มาก — ลดแล้วได้ volume คุ้ม)
- ลดราคาน้อยลง (เพราะ demand ไม่ sensitive — ลดเกินแค่กินทุน)
- หยุดโปร (เพราะ inelastic — ลูกค้าซื้ออยู่แล้ว ไม่ต้องลด)


### 1.4 Deliverables & Validation Plan

**สิ่งที่จะส่งมอบให้ลูกค้า:**
1. `Promotion Recommendation Table` — output หลักที่ใช้ตัดสินใจ
2. `Elasticity Curve` ต่อ product category — อธิบายว่าทำไมถึงแนะนำแบบนั้น
3. `Before vs After Simulation` — แสดง revenue/profit ถ้าใช้ optimized discount vs random discount

**วิธี Validate ว่า Idea นี้ Work (โดยไม่ต้อง deploy จริง):**
- **Baseline:** สมมติว่าเดิมร้านลด 20% ทุกสินค้าเหมือนกัน (random strategy)
- **Model:** ใช้ elasticity แนะนำ discount % ที่ต่างกันต่อสินค้า
- **Compare:** Simulate revenue และ profit บน historical mock data
- ถ้า optimized > baseline → idea validated ✅

**Assumption หลัก:**
- ข้อมูล mock มาจาก 1 ปี (2024) ใช้แทน historical data จริง
- cogs ที่ mock ไว้ใช้เป็น proxy ต้นทุนจริง

## STEP 3 — Data Cleaning & Preparation
---

แต่ละปัญหาที่เจอใน Step 2 จะแก้อย่างเป็นระบบ พร้อม log ว่าลบ/แก้ไปกี่ rows


```
import os
os.makedirs('./output', exist_ok=True)

# ── Helper: log cleaning action ────────────────────────
cleaning_log = []

def log_clean(table, issue, before, after, action):
    removed = before - after
    cleaning_log.append({
        'Table': table, 'Issue': issue,
        'Before': before, 'Removed/Fixed': removed, 'After': after, 'Action': action
    })
    print(f"  [{table}] {issue}: {before} → {after} (fixed {removed})")

print("Cleaning log helper ready ✅")
```

### 3.1 Clean: Sales Transaction

```
sales_clean = sales.copy()
n0 = len(sales_clean)
print(f"Starting rows: {n0}")

# ── 1. Remove duplicate rows ────────────────────────────
sales_clean = sales_clean.drop_duplicates()
log_clean("Sales", "Duplicate rows", n0, len(sales_clean), "drop_duplicates()")

# ── 2. Standardize datetime (2 formats exist) ───────────
def parse_datetime(s):
    for fmt in ('%Y-%m-%d %H:%M:%S', '%d-%m-%Y %H:%M', '%Y-%m-%d %H:%M'):
        try:
            return pd.to_datetime(s, format=fmt)
        except:
            pass
    return pd.NaT

n_before = len(sales_clean)
sales_clean['datetime'] = sales_clean['datetime'].apply(parse_datetime)
n_null_dt = sales_clean['datetime'].isna().sum()
print(f"  [Sales] datetime parse failures: {n_null_dt}")

# ── 3. Drop future datetime (year > 2024) ───────────────
before = len(sales_clean)
sales_clean = sales_clean[sales_clean['datetime'].dt.year <= 2024]
log_clean("Sales", "Future datetime", before, len(sales_clean), "drop rows where year > 2024")

# ── 4. Convert price & qty to numeric ───────────────────
sales_clean['price'] = pd.to_numeric(sales_clean['price'], errors='coerce')
sales_clean['qty']   = pd.to_numeric(sales_clean['qty'],   errors='coerce')

# ── 5. Drop rows where price is null or 0 ───────────────
before = len(sales_clean)
sales_clean = sales_clean[(sales_clean['price'].notna()) & (sales_clean['price'] > 0)]
log_clean("Sales", "Price null or 0", before, len(sales_clean), "drop rows")

# ── 6. Drop rows with negative qty (untracked returns) ──
before = len(sales_clean)
sales_clean = sales_clean[sales_clean['qty'] > 0]
log_clean("Sales", "Negative qty", before, len(sales_clean), "drop rows (untracked returns)")

# ── 7. Flag outlier qty (> 100) — keep but flag ─────────
sales_clean['is_bulk_order'] = sales_clean['qty'] > 100
n_bulk = sales_clean['is_bulk_order'].sum()
print(f"  [Sales] Outlier qty (>100): {n_bulk} flagged as bulk_order (kept)")

# ── 8. Compute revenue column ───────────────────────────
sales_clean['revenue'] = sales_clean['price'] * sales_clean['qty']

print(f"\nFinal Sales rows: {len(sales_clean)} (started: {n0})")
```

### 3.2 Clean: Product Master

```
prod_clean = prod.copy()

# ── 1. Remove duplicates ────────────────────────────────
before = len(prod_clean)
prod_clean = prod_clean.drop_duplicates(subset='product_id', keep='first')
log_clean("Product", "Duplicate product_id", before, len(prod_clean), "keep first")

# ── 2. Clean price column (strip " บาท" and cast) ───────
prod_clean['price'] = (prod_clean['price']
                       .astype(str)
                       .str.replace('บาท', '', regex=False)
                       .str.strip())
prod_clean['price'] = pd.to_numeric(prod_clean['price'], errors='coerce')
n_bad_price = prod_clean['price'].isna().sum()
print(f"  [Product] price parse failures after strip: {n_bad_price}")

# ── 3. Standardize product_taxonomies ───────────────────
category_map = {
    'Beverage'    : 'เครื่องดื่ม',
    'เครื่องดื่ม ': 'เครื่องดื่ม',  # trailing space
    'เครืองดืม'  : 'เครื่องดื่ม',   # missing tone mark
}
prod_clean['product_taxonomies'] = (prod_clean['product_taxonomies']
                                    .str.strip()
                                    .replace(category_map))

# ── 4. Fill missing category with 'Unknown' ─────────────
before_null = prod_clean['product_taxonomies'].isna().sum()
prod_clean['product_taxonomies'] = prod_clean['product_taxonomies'].fillna('Unknown')
print(f"  [Product] Missing category filled with 'Unknown': {before_null} rows")

print(f"\nCategories after cleaning:")
print(prod_clean['product_taxonomies'].value_counts().to_string())
```

### 3.3 Clean: Promotion Master

```
promo_clean = promo.copy()

# ── 1. Standardize discount: if > 1 → divide by 100 ────
promo_clean['discount'] = pd.to_numeric(promo_clean['discount'], errors='coerce')
mask_pct = promo_clean['discount'] > 1
n_fixed = mask_pct.sum()
promo_clean.loc[mask_pct, 'discount'] = promo_clean.loc[mask_pct, 'discount'] / 100
log_clean("Promotion", "Discount as integer %", len(promo_clean), len(promo_clean), f"divide by 100 ({n_fixed} rows)")

# ── 2. Fill missing discount with median ────────────────
median_disc = promo_clean['discount'].median()
n_null_disc = promo_clean['discount'].isna().sum()
promo_clean['discount'] = promo_clean['discount'].fillna(median_disc)
print(f"  [Promo] Missing discount filled with median ({median_disc:.2f}): {n_null_disc} rows")

# ── 3. Parse dates (2 formats) ──────────────────────────
for col in ['start_date', 'end_date']:
    promo_clean[col] = pd.to_datetime(promo_clean[col], 
                                       format='mixed', dayfirst=False, errors='coerce')

# ── 4. Drop rows where end_date < start_date ────────────
before = len(promo_clean)
promo_clean = promo_clean[promo_clean['end_date'] >= promo_clean['start_date']]
log_clean("Promotion", "end_date < start_date", before, len(promo_clean), "drop rows")

print(f"\nPromotion Master after cleaning: {len(promo_clean)} rows")
```

### 3.4 Clean: Customer & Store Master

```
# ── Customer Master ────────────────────────────────────
cus_clean = cus.copy()

before = len(cus_clean)
cus_clean = cus_clean.drop_duplicates(subset='customer_id', keep='first')
log_clean("Customer", "Duplicate customer_id", before, len(cus_clean), "keep first")

segment_map = {
    'Loyal': 'loyal', 'LOYAL': 'loyal', 'loyl': 'loyal',
    'ocasional': 'occasional', 'Occasional': 'occasional',
}
cus_clean['customer_taxonomies'] = (cus_clean['customer_taxonomies']
                                    .replace(segment_map))
cus_clean['customer_taxonomies'] = cus_clean['customer_taxonomies'].fillna('unknown')
print("Customer segments after cleaning:")
print(cus_clean['customer_taxonomies'].value_counts().to_string())

# ── Store Master ────────────────────────────────────────
store_clean = store.copy()

store_type_map = {
    'Convenience': 'convenience', 'CONVENIENCE': 'convenience',
    'convenience store': 'convenience',
}
store_clean['store_taxonomies'] = (store_clean['store_taxonomies']
                                   .replace(store_type_map))
store_clean['region'] = store_clean['region'].fillna('Unknown')
print("\nStore types after cleaning:")
print(store_clean['store_taxonomies'].value_counts().to_string())
```

### 3.5 Cleaning Summary

```
log_df = pd.DataFrame(cleaning_log)
print("=== Cleaning Log ===\n")
print(log_df.to_string(index=False))
```

### 3.6 Build Master Dataset (Join All Tables)

```
# ── Explode promotion → one row per product ────────────
promo_exploded = promo_clean.copy()
promo_exploded['product_id'] = promo_exploded['product_id'].str.split(',')
promo_exploded = promo_exploded.explode('product_id')
promo_exploded['product_id'] = promo_exploded['product_id'].str.strip()

# ── Join Sales + Product ────────────────────────────────
df = sales_clean.merge(prod_clean[['product_id','product_taxonomies','cogs']],
                       on='product_id', how='left')

# ── Join Sales + Store ──────────────────────────────────
df = df.merge(store_clean[['store_id','store_taxonomies','region']],
              on='store_id', how='left')

# ── Join Sales + Customer ───────────────────────────────
df = df.merge(cus_clean[['customer_id','customer_taxonomies','channel']],
              on='customer_id', how='left')

# ── Join Sales + Promotion ──────────────────────────────
df = df.merge(
    promo_exploded[['promotion_id','discount','promotion_type']],
    on='promotion_id', how='left'
)

# ── Derived columns ─────────────────────────────────────
df['discount']         = df['discount'].fillna(0)       # no promotion = 0% discount
df['effective_price']  = df['price'] * (1 - df['discount'])
df['gross_profit']     = (df['effective_price'] - df['cogs']) * df['qty']
df['profit_margin_pct']= (df['gross_profit'] / (df['effective_price'] * df['qty'])) * 100
df['has_promotion']    = df['promotion_id'].notna()
df['month']            = df['datetime'].dt.month
df['dayofweek']        = df['datetime'].dt.day_name()

print(f"Master dataframe: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head(3)
```

## STEP 4 — Model Selection & Experimenting
---

### Approach: Price Elasticity of Demand (Log-Log Regression)

**ทำไมใช้ Log-Log Regression?**
- เป็น econometric standard สำหรับ price elasticity
- Coefficient ที่ได้คือ elasticity โดยตรง: ถ้า β = -1.5 แปลว่า ราคาเพิ่ม 1% → demand ลด 1.5%
- ตีความได้ง่าย เหมาะสำหรับ present ให้ business owner เข้าใจ
- ไม่ซับซ้อนเกินไป เหมาะกับ mock data ที่มี

**ขั้นตอน:**
1. คำนวณ elasticity ต่อ product category
2. จำแนกสินค้าเป็น elastic / inelastic / unit elastic
3. หา optimal discount % ที่ maximize gross profit


```
from scipy import stats

# ── Aggregate to product-level for elasticity calc ─────
# ใช้ distinct price points ต่อ product (ได้จาก transactions ที่ discount ต่างกัน)
price_qty = (df.groupby(['product_id', 'product_taxonomies', 'effective_price'])
               .agg(total_qty=('qty','sum'),
                    avg_cogs=('cogs','mean'),
                    transaction_count=('po_id','count'))
               .reset_index())

# กรอง: ต้องมีอย่างน้อย 5 price points ถึงจะ estimate elasticity ได้
price_qty_count = price_qty.groupby('product_id')['effective_price'].nunique()
valid_products  = price_qty_count[price_qty_count >= 5].index

price_qty_valid = price_qty[price_qty['product_id'].isin(valid_products)].copy()
price_qty_valid = price_qty_valid[
    (price_qty_valid['effective_price'] > 0) & (price_qty_valid['total_qty'] > 0)
]

print(f"Products with enough price variation: {len(valid_products)}")
print(f"Rows for regression: {len(price_qty_valid)}")
```

```
# ── Run Log-Log Regression per product ─────────────────
elasticity_results = []

for pid in valid_products:
    subset = price_qty_valid[price_qty_valid['product_id'] == pid].copy()
    if len(subset) < 5:
        continue

    log_p = np.log(subset['effective_price'].values)
    log_q = np.log(subset['total_qty'].values)

    slope, intercept, r_value, p_value, std_err = stats.linregress(log_p, log_q)

    # slope = price elasticity (should be negative for normal goods)
    elasticity_results.append({
        'product_id'         : pid,
        'product_taxonomies' : subset['product_taxonomies'].iloc[0],
        'elasticity'         : round(slope, 3),
        'r_squared'          : round(r_value**2, 3),
        'p_value'            : round(p_value, 4),
        'avg_price'          : round(subset['effective_price'].mean(), 2),
        'avg_cogs'           : round(subset['avg_cogs'].mean(), 2),
    })

elast_df = pd.DataFrame(elasticity_results)

# Classify elasticity type
def classify(e):
    if e < -1:   return 'elastic'       # ลดราคาแล้ว demand เพิ่มมาก → คุ้มที่จะลด
    elif e > -1: return 'inelastic'     # ลดราคาแล้ว demand แทบไม่เพิ่ม → ไม่ควรลด
    else:        return 'unit_elastic'

elast_df['elasticity_type'] = elast_df['elasticity'].apply(classify)

print(f"Elasticity estimated for {len(elast_df)} products")
print("\nElasticity type distribution:")
print(elast_df['elasticity_type'].value_counts().to_string())
print("\nSample results:")
print(elast_df.head(8).to_string(index=False))
```

```
# ── Visualize Elasticity Distribution ──────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: elasticity distribution
colors_map = {'elastic': '#4C8BF5', 'inelastic': '#E8555A', 'unit_elastic': '#FBBC04'}
type_counts = elast_df['elasticity_type'].value_counts()

axes[0].bar(type_counts.index,
            type_counts.values,
            color=[colors_map.get(t,'gray') for t in type_counts.index],
            edgecolor='white', width=0.5)
axes[0].set_title('Product Elasticity Classification', fontsize=13)
axes[0].set_ylabel('Number of Products')
axes[0].axhline(0, color='black', linewidth=0.8)

# Plot 2: elasticity by category (boxplot)
cat_data = [elast_df[elast_df['product_taxonomies']==cat]['elasticity'].values
            for cat in elast_df['product_taxonomies'].unique()]
cat_labels = elast_df['product_taxonomies'].unique()

axes[1].boxplot(cat_data, labels=cat_labels, patch_artist=True,
                boxprops=dict(facecolor='#D6E4F7', color='#2C5F9E'),
                medianprops=dict(color='#E8555A', linewidth=2))
axes[1].axhline(-1, color='orange', linestyle='--', linewidth=1.5, label='Elasticity = -1 (unit elastic)')
axes[1].set_title('Elasticity Range by Product Category', fontsize=13)
axes[1].set_ylabel('Price Elasticity')
axes[1].tick_params(axis='x', rotation=25)
axes[1].legend()

plt.suptitle('Step 4 — Price Elasticity Analysis', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('./output/fig_03_elasticity.png', dpi=120, bbox_inches='tight')
plt.show()
print("Fig 03 saved ✅")
```

### 4.2 Optimal Discount Calculation

สำหรับแต่ละสินค้า หา discount % ที่ทำให้ **Gross Profit สูงสุด**  
โดยใช้ elasticity formula:  
`Δ Qty % ≈ elasticity × Δ Price %`  
`Gross Profit = (effective_price − cogs) × estimated_qty`


```
from scipy.optimize import minimize_scalar

def estimate_gross_profit(discount_pct, base_price, cogs, base_qty, elasticity):
    """
    ประมาณ gross profit ที่ discount level หนึ่งๆ
    ใช้ point elasticity: %Δqty = elasticity × %Δprice
    """
    price_change_pct = -discount_pct                        # ลดราคา d% → price_change = -d%
    qty_change_pct   = elasticity * price_change_pct        # %Δqty = e × %Δprice
    new_qty          = base_qty * (1 + qty_change_pct / 100)
    effective_price  = base_price * (1 - discount_pct / 100)
    gross_profit     = (effective_price - cogs) * max(new_qty, 0)
    return gross_profit

# ── Compute recommendation per product ─────────────────
recommendations = []
discount_range = np.arange(0, 41, 1)   # 0% to 40% in steps of 1%

for _, row in elast_df.iterrows():
    pid         = row['product_id']
    base_price  = row['avg_price']
    cogs        = row['avg_cogs']
    elasticity  = row['elasticity']
    elast_type  = row['elasticity_type']

    # Base qty from clean sales
    base_qty = df[df['product_id'] == pid]['qty'].sum()
    if base_qty == 0:
        continue

    # Current avg discount this product gets
    current_disc = df[df['product_id'] == pid]['discount'].mean() * 100

    # Find optimal discount
    profits = [estimate_gross_profit(d, base_price, cogs, base_qty, elasticity)
               for d in discount_range]
    optimal_idx     = int(np.argmax(profits))
    optimal_disc    = discount_range[optimal_idx]
    current_profit  = estimate_gross_profit(current_disc, base_price, cogs, base_qty, elasticity)
    optimal_profit  = profits[optimal_idx]

    # Expected revenue change
    price_chg_pct    = -optimal_disc
    qty_chg_pct      = elasticity * price_chg_pct
    exp_rev_change   = round(qty_chg_pct - optimal_disc, 1)   # simple estimate
    exp_profit_change= round((optimal_profit - current_profit) / (abs(current_profit) + 1e-9) * 100, 1)

    # Determine action
    if optimal_disc < current_disc - 2:
        action = "Reduce Discount"
    elif optimal_disc > current_disc + 2:
        action = "Increase Discount"
    elif optimal_disc == 0:
        action = "Stop Promotion"
    else:
        action = "Keep Current"

    recommendations.append({
        'product_id'           : pid,
        'product_taxonomies'   : row['product_taxonomies'],
        'elasticity'           : elasticity,
        'elasticity_type'      : elast_type,
        'current_discount_pct' : round(current_disc, 1),
        'recommended_discount_pct': optimal_disc,
        'expected_revenue_change_pct': exp_rev_change,
        'expected_profit_change_pct' : exp_profit_change,
        'action'               : action,
    })

rec_df = pd.DataFrame(recommendations)
print(f"Recommendations generated: {len(rec_df)} products")
print("\nAction distribution:")
print(rec_df['action'].value_counts().to_string())
```

### 5.2 Baseline Comparison — Validate ว่า Model ดีกว่า Random มั้ย

**วิธีคิด:**
- **Baseline:** สมมติร้านลด 15% ทุกสินค้าเหมือนกัน (วิธีที่เจ้าของร้านทำอยู่)
- **Model:** ใช้ recommended_discount_pct ต่อสินค้า
- **Compare:** total estimated gross profit ระหว่าง 2 strategy


```
# Simulate baseline vs model on the product set
baseline_profits = []
model_profits    = []

for _, row in rec_df.iterrows():
    pid        = row['product_id']
    base_price = elast_df[elast_df['product_id']==pid]['avg_price'].values[0]
    cogs       = elast_df[elast_df['product_id']==pid]['avg_cogs'].values[0]
    elasticity = row['elasticity']
    base_qty   = df[df['product_id']==pid]['qty'].sum()

    baseline_p = estimate_gross_profit(15, base_price, cogs, base_qty, elasticity)
    model_p    = estimate_gross_profit(
                     row['recommended_discount_pct'],
                     base_price, cogs, base_qty, elasticity)

    baseline_profits.append(baseline_p)
    model_profits.append(model_p)

total_baseline = sum(baseline_profits)
total_model    = sum(model_profits)
improvement    = (total_model - total_baseline) / abs(total_baseline) * 100

print("=" * 50)
print("BASELINE vs MODEL — Gross Profit Comparison")
print("=" * 50)
print(f"  Baseline (flat 15% discount) : {total_baseline:>12,.0f} THB")
print(f"  Model (optimized discount)   : {total_model:>12,.0f} THB")
print(f"  Improvement                  : {improvement:>+11.1f}%")
print("=" * 50)
print("\n→ If improvement > 0, the elasticity model adds value ✅")
```

```
# ── Visualize Baseline vs Model ────────────────────────
labels   = ['Baseline\n(Flat 15% Discount)', 'Model\n(Elasticity-Optimized)']
values   = [total_baseline, total_model]
bar_cols = ['#E8E8E8', '#4C8BF5']

plt.figure(figsize=(7, 5))
bars = plt.bar(labels, values, color=bar_cols, edgecolor='white', width=0.45)
plt.ylabel('Estimated Total Gross Profit (THB)')
plt.title('Baseline vs Optimized Promotion Strategy\n(Simulated on Mock Data)', fontsize=13)

for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() * 1.01,
             f'{val:,.0f}', ha='center', fontsize=11, fontweight='bold')

diff_pct_label = f'+{improvement:.1f}%' if improvement >= 0 else f'{improvement:.1f}%'
plt.annotate(f'Improvement: {diff_pct_label}',
             xy=(1, total_model), xytext=(0.5, max(values)*0.6),
             fontsize=12, color='#2C5F9E', fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='#2C5F9E'))

plt.tight_layout()
plt.savefig('./output/fig_05_baseline_vs_model.png', dpi=120, bbox_inches='tight')
plt.show()
print("Fig 05 saved ✅")
```

### 5.3 KPI Summary

```
# ── KPI 1: Profit Margin % ─────────────────────────────
valid_margin = df[(df['profit_margin_pct'].notna()) &
                  (np.isfinite(df['profit_margin_pct']))]
avg_margin_with_promo    = valid_margin[valid_margin['has_promotion']]['profit_margin_pct'].mean()
avg_margin_without_promo = valid_margin[~valid_margin['has_promotion']]['profit_margin_pct'].mean()

# ── KPI 2: Promotion ROI ────────────────────────────────
df['discount_cost'] = df['price'] * df['discount'] * df['qty']
promo_transactions  = df[df['has_promotion'] & (df['discount_cost'] > 0)]
total_discount_cost = promo_transactions['discount_cost'].sum()
total_promo_revenue = promo_transactions['revenue'].sum()
promotion_roi       = (total_promo_revenue - total_discount_cost) / total_discount_cost * 100

# ── KPI 3: Promotion Uplift % ──────────────────────────
baseline_rev_per_tx = df[~df['has_promotion']]['revenue'].mean()
promo_rev_per_tx    = df[df['has_promotion']]['revenue'].mean()
uplift_pct          = (promo_rev_per_tx - baseline_rev_per_tx) / baseline_rev_per_tx * 100

print("=" * 55)
print("KPI SUMMARY")
print("=" * 55)
print(f"  KPI 1 | Profit Margin % (with promo)    : {avg_margin_with_promo:.1f}%")
print(f"         Profit Margin % (without promo)  : {avg_margin_without_promo:.1f}%")
print(f"         Delta                            : {avg_margin_with_promo - avg_margin_without_promo:+.1f}%")
print()
print(f"  KPI 2 | Promotion ROI                   : {promotion_roi:.1f}%")
print()
print(f"  KPI 3 | Promotion Revenue Uplift        : {uplift_pct:+.1f}%")
print("=" * 55)
```

### 6.3 Assumption & Limitation

|---|---|
| **Mock Data** | ใช้ข้อมูลสังเคราะห์แทน transaction จริง — elasticity ที่ได้เป็น simulation เท่านั้น |
