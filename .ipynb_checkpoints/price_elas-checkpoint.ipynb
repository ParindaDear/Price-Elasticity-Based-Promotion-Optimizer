## STEP 1: Business Understanding

### ปัญหาที่แท้จริงคืออะไร?

ร้าน Retail ขนาดเล็ก-กลาง (SME) มักจัดโปรโมชั่นแบบ **"จัดไปก่อน แล้วค่อยดูผล"**  
ซึ่งทำให้:
- บางสินค้าลด 20% แต่ยอดขายไม่ขึ้นเลย → Margin หาย เปล่าประโยชน์
- บางสินค้าลดแค่ 5% ก็ขายดีขึ้นมาก → แต่กลับไม่จัดโปรโมชั่น เสียโอกาส

### Solution ที่เสนอ: Price Elasticity-Based Optimizer

**ทำอะไร?**  
ประมาณค่า Price Elasticity of Demand (PED) ของแต่ละสินค้า จากข้อมูลขายจริง  
แล้วหาส่วนลดที่ทำให้ Revenue หรือ Gross Profit สูงสุด

```
Input:  Sales Transaction + Product Master + Promotion Master + Customer + Store
Output: Promotion Recommendation Table (ลดเท่าไหร่ → คาด Revenue/Profit เปลี่ยนแค่ไหน)
```

**ทำไม Rule-based ธรรมดาไม่พอ?**
- Rule-based: "ถ้าสต็อกเยอะ → ลด 20%" → ไม่ได้มองพฤติกรรมลูกค้าจริง
- Dashboard: บอกว่า "ยอดขายลด" แต่ไม่บอกว่า "ควรทำอะไร"
- Elasticity Model: **บอกได้ว่าลดเท่าไหร่ถึงคุ้ม** และผลต่อ Margin เป็นเท่าไหร่

### Stakeholders
- **Primary User:** เจ้าของร้าน / ผู้จัดการฝ่ายการตลาด
- **Decision:** จะจัดโปรโมชั่นสินค้าไหน ลดเท่าไหร่ ช่วงไหน

### KPI Definitions
| KPI | นิยาม | เป้าหมาย |
|-----|--------|----------|
| Gross Profit Margin | (Revenue - COGS) / Revenue | เพิ่มขึ้น vs baseline |
| Promotion ROI | Revenue Uplift / Discount Cost | > 1.0 (คุ้มทุน) |
| Revenue Uplift | % เปลี่ยนแปลงของยอดขาย vs ก่อนโปร | บวกและมีนัยสำคัญ |

## STEP 3: Data Cleaning & Preparation

````python
# ========== 3.1 Clean product_master ==========

# 3.1.1 Normalize product_taxonomies: ลบ space หน้า-หลัง และแปลงเป็นตัวพิมพ์เล็กทั้งหมด
# (แก้ปัญหา 'beverage' / 'Beverage' / ' beverage' ที่จริงๆควรเป็น category เดียวกัน)
product['product_taxonomies'] = (
    product['product_taxonomies']
    .str.strip()
    .str.lower()
)

# 3.1.2 Strip 'baht' และ whitespace จาก price แล้วแปลงเป็น float
product['price'] = (
    product['price']
    .str.replace(r'[^\d.]', '', regex=True)  # เอาเฉพาะตัวเลขและจุด
    .astype(float)
)

# 3.1.3 Fill null product_taxonomies เป็น 'unknown'
product['product_taxonomies'] = product['product_taxonomies'].fillna('unknown')

# 3.1.4 Drop duplicate rows (เจอ 8 แถวที่ซ้ำกันทุก column)
print(f'Duplicate rows found: {product.duplicated().sum()}')
product = product.drop_duplicates().reset_index(drop=True)

print('\nproduct_master cleaned:')
print(f'Rows: {len(product)}')
print(f'Null check:\n{product.isnull().sum()}')
print('\nUnique product_taxonomies after cleaning:')
print(product['product_taxonomies'].value_counts())
````

````python
# ========== 3.2 Clean promotion_master ==========

# 3.2.1 แปลง start_date และ end_date ให้เป็น datetime dtype (เดิมเป็น object/string)
for col in ['start_date', 'end_date']:
    raw_col = promo[col].astype(str)
    parsed  = pd.to_datetime(raw_col, errors='coerce')

    # บางแถวอาจเป็น format DD/MM/YYYY (เช่น 18/04/2024) ลอง parse อีกครั้งด้วย dayfirst=True
    nat_mask = parsed.isna()
    if nat_mask.any():
        parsed.loc[nat_mask] = pd.to_datetime(raw_col.loc[nat_mask], dayfirst=True, errors='coerce')

    promo[col] = parsed

# 3.2.2 buy1get1 ไม่มี discount → effective discount ≈ 0.50
buy1get1_mask = (promo['promotion_type'] == 'buy1get1') & (promo['discount'].isna())
promo.loc[buy1get1_mask, 'discount'] = 0.50

# percent_off ที่ null → conservative fill = 0.10
promo['discount'] = promo['discount'].fillna(0.10)

# 3.2.3 Normalize discount: ถ้า > 1 แสดงว่าเป็น % (เช่น 19 หมายถึง 19%)
promo.loc[promo['discount'] > 1, 'discount'] = promo.loc[promo['discount'] > 1, 'discount'] / 100

print('promotion_master cleaned:')
print(promo[['promotion_id', 'discount', 'start_date', 'end_date', 'promotion_type']].dtypes)
print()
print(promo[['promotion_id', 'discount', 'start_date', 'end_date', 'promotion_type']].head(10))
````

````python
# ========== 3.3 Clean sales_transaction ==========

# 3.3.1 Drop duplicate rows (เจอ 50 แถวที่ซ้ำกันทุก column)
print(f'Duplicate rows found: {sales.duplicated().sum()}')
sales = sales.drop_duplicates().reset_index(drop=True)

# 3.3.2 Parse datetime ให้เป็น datetime dtype (เดิมเป็น object)
sales['datetime'] = pd.to_datetime(sales['datetime'], errors='coerce')

# 3.3.3 กรอง price null ออก
print(f'Rows with null price: {sales["price"].isna().sum()}')
sales = sales[sales['price'].notna()].copy()

# 3.3.4 กรอง returns (qty ติดลบ) ออก
print(f'Rows with negative qty (returns): {(sales["qty"] < 0).sum()}')
sales = sales[sales['qty'] > 0].copy()

# 3.3.5 ลบ qty ที่เป็น outlier ผิดปกติ (500, 888, 999) ออก
outlier_qty_values = [500, 888, 999]
print(f'Rows with qty outliers {outlier_qty_values}: {sales["qty"].isin(outlier_qty_values).sum()}')
sales = sales[~sales['qty'].isin(outlier_qty_values)].copy()

# 3.3.6 สร้าง revenue column
sales_clean = sales.copy()
sales_clean['revenue'] = sales_clean['price'] * sales_clean['qty']

# 3.3.7 สร้าง flag ว่า transaction นี้มีโปรโมชั่นหรือเปล่า
sales_clean['has_promotion'] = sales_clean['promotion_id'].notna().astype(int)

print(f'\nClean sales rows: {len(sales_clean):,}')
print(f'Date range: {sales_clean["datetime"].min()} to {sales_clean["datetime"].max()}')
print('\nQty stats after cleaning:')
print(sales_clean['qty'].describe().round(2))
````

````python
# ========== 3.4 Clean store_master ==========

# 3.4.1 Normalize store_taxonomies: ลบ space หน้า-หลัง, แปลงเป็นตัวพิมพ์เล็กทั้งหมด,
#       และตัดคำว่า 'store' ที่ติดมาด้านหลังออก
#       (รวม 'convenience', 'CONVENIENCE', 'Convenience', 'convenience store'
#        ให้กลายเป็น category เดียวกันคือ 'convenience')
store['store_taxonomies'] = (
    store['store_taxonomies']
    .str.strip()
    .str.lower()
    .str.replace(r'\s+store$', '', regex=True)
)

print('store_master cleaned:')
print('\nUnique store_taxonomies after cleaning:')
print(store['store_taxonomies'].value_counts())
````

````python
# ========== 3.5 Clean customer_master ==========

# 3.5.1 Normalize customer_taxonomies: ลบ space หน้า-หลัง และแปลงเป็นตัวพิมพ์เล็กทั้งหมด
# (แก้ปัญหา 'loyal' / 'LOYAL' / 'Loyal' ที่จริงๆควรเป็น category เดียวกัน)
customer['customer_taxonomies'] = (
    customer['customer_taxonomies']
    .str.strip()
    .str.lower()
)

# 3.5.2 แก้ typo: 'ocasional' -> 'occasional'
# (ระหว่างทางเจอ 'loyl' ด้วย ซึ่งเป็น typo ของ 'loyal' เลยแก้พร้อมกัน)
typo_map = {'ocasional': 'occasional', 'loyl': 'loyal'}
customer['customer_taxonomies'] = customer['customer_taxonomies'].replace(typo_map)

# 3.5.3 Fill null customer_taxonomies เป็น 'unknown'
customer['customer_taxonomies'] = customer['customer_taxonomies'].fillna('unknown')

# 3.5.4 Drop duplicate rows (เจอ 15 แถวที่ซ้ำกันทุก column)
print(f'Duplicate rows found: {customer.duplicated().sum()}')
customer = customer.drop_duplicates().reset_index(drop=True)

print('\ncustomer_master cleaned:')
print(f'Rows: {len(customer)}')
print('\nUnique customer_taxonomies after cleaning:')
print(customer['customer_taxonomies'].value_counts())
````

````python
# ========== 3.6 Merge tables ==========

# Step 1: Join sales → product (ได้ COGS, product category)
df = sales_clean.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id',
    how='left',
    suffixes=('_sold', '_list')  # price_sold = ราคาจริง, price_list = ราคาตั้ง
)

# Step 2: Join → promotion (ได้ discount rate)
promo_slim = promo[['promotion_id', 'discount', 'promotion_type', 'start_date', 'end_date']]
df = df.merge(promo_slim, on='promotion_id', how='left')

# Step 3: Join → customer
df = df.merge(
    customer[['customer_id', 'customer_taxonomies', 'age_group', 'channel']],
    on='customer_id',
    how='left'
)

# Step 4: Join → store
df = df.merge(
    store[['store_id', 'store_taxonomies', 'region', 'store_size_sqm']],
    on='store_id',
    how='left'
)

# คำนวณ effective discount rate (ถ้าไม่มีโปร = 0)
df['discount'] = df['discount'].fillna(0.0)

# คำนวณ effective price (ราคาที่ลูกค้าจ่ายจริง)
df['effective_price'] = df['price_list'] * (1 - df['discount'])

# คำนวณ gross profit
df['gross_profit'] = (df['price_sold'] - df['cogs']) * df['qty']

print('Merged DataFrame shape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nSample:')
df[['product_id', 'price_sold', 'price_list', 'discount', 'effective_price', 'qty', 'revenue', 'gross_profit']].head(5)
````

````python
# ========== EDA Post-Merge: Revenue by Category ==========

rev_by_cat = (
    df.groupby('product_taxonomies')['revenue']
    .sum()
    .sort_values(ascending=True)
)

fig, ax = plt.subplots(figsize=(9, 5))
rev_by_cat.plot(kind='barh', ax=ax, color='steelblue')
ax.set_title('Total Revenue by Product Category')
ax.set_xlabel('Revenue (Baht)')
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.tight_layout()
plt.show()
````

````python
# ========== EDA: Discount vs Avg Qty Sold ==========
# สังเกตว่าส่วนลดสัมพันธ์กับ qty อย่างไร

# Bin discount เป็นช่วง
df['discount_bin'] = pd.cut(
    df['discount'],
    bins=[-0.01, 0, 0.10, 0.20, 0.30, 0.50, 1.0],
    labels=['0%', '1-10%', '11-20%', '21-30%', '31-50%', '>50%']
)

avg_qty = df.groupby('discount_bin', observed=True)['qty'].mean().reset_index()

fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(avg_qty['discount_bin'], avg_qty['qty'], color='coral', edgecolor='white')
ax.set_title('Average Qty Sold by Discount Level')
ax.set_xlabel('Discount Range')
ax.set_ylabel('Avg Qty')
plt.tight_layout()
plt.show()

print(avg_qty)
````

## STEP 4: Model Selection & Experimenting

### แนวคิด: Price Elasticity of Demand (PED)

$$PED = \frac{\% \Delta Quantity}{\% \Delta Price}$$

- **PED < -1** → Elastic (ลดราคาแล้วยอดขายขึ้นมาก → ควรจัดโปร)
- **-1 < PED < 0** → Inelastic (ลดราคาแล้วยอดขายขึ้นน้อย → ไม่ควรลดมาก)
- **PED ≈ 0** → ลูกค้าไม่สนใจราคา

### วิธีประมาณ Elasticity จากข้อมูลที่มี

เราใช้ **Log-Log OLS Regression** ระดับสินค้า:

$$\ln(Qty) = \alpha + \beta_1 \cdot \ln(Price) + \epsilon$$

- $\beta_1$ คือค่า Elasticity โดยตรง
- ใช้ข้อมูลหลายๆ transaction ของสินค้าเดียวกัน (ราคาต่างกันตาม promotion)
- สินค้าที่มี transaction น้อยกว่า 10 ครั้ง → ใช้ค่า median ของ category แทน

````python
# ========== 4.1 Compute Price Elasticity per Product ==========
from scipy import stats

def compute_elasticity(group):
    """Log-Log OLS regression: ln(qty) ~ ln(price)"""
    if len(group) < 5:
        return np.nan  # ข้อมูลน้อยเกินไป
    
    log_price = np.log(group['price_sold'])
    log_qty   = np.log(group['qty'])
    
    # กรอง inf หรือ nan
    mask = np.isfinite(log_price) & np.isfinite(log_qty)
    if mask.sum() < 5:
        return np.nan
    
    slope, intercept, r_value, p_value, std_err = stats.linregress(
        log_price[mask], log_qty[mask]
    )
    return slope  # นี่คือ elasticity coefficient


# คำนวณ elasticity ทุก product
elasticity_by_product = (
    df.groupby('product_id')
    .apply(compute_elasticity)
    .reset_index()
    .rename(columns={0: 'elasticity'})
)

# Join กับ product info
elasticity_df = elasticity_by_product.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id',
    how='left'
)

# สินค้าที่ elasticity เป็น NaN → fill ด้วย median ของ category
category_median = elasticity_df.groupby('product_taxonomies')['elasticity'].median()
elasticity_df['elasticity'] = elasticity_df.apply(
    lambda row: category_median.get(row['product_taxonomies'], -1.0)
    if pd.isna(row['elasticity']) else row['elasticity'],
    axis=1
)

print(f'Elasticity computed for {elasticity_df["elasticity"].notna().sum()} products')
print('\nElasticity distribution:')
print(elasticity_df['elasticity'].describe().round(3))
````

````python
# ========== 4.2 Visualize Elasticity Distribution ==========

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Histogram
axes[0].hist(elasticity_df['elasticity'].dropna(), bins=25, color='steelblue', edgecolor='white')
axes[0].axvline(-1, color='red', linestyle='--', label='Elasticity = -1 (unit elastic)')
axes[0].axvline(0,  color='orange', linestyle='--', label='Elasticity = 0')
axes[0].set_title('Price Elasticity Distribution across Products')
axes[0].set_xlabel('Elasticity Coefficient')
axes[0].set_ylabel('Number of Products')
axes[0].legend(fontsize=8)

# By category
cat_elast = elasticity_df.groupby('product_taxonomies')['elasticity'].mean().sort_values()
colors = ['#d32f2f' if e > -1 else '#388e3c' for e in cat_elast]
axes[1].barh(cat_elast.index, cat_elast.values, color=colors)
axes[1].axvline(-1, color='black', linestyle='--', linewidth=0.8)
axes[1].set_title('Avg Elasticity by Product Category\n(Green = Elastic = ควรจัดโปร)')
axes[1].set_xlabel('Avg Elasticity')

plt.tight_layout()
plt.show()
````

````python
# ========== 4.3 Optimal Discount Finder ==========

def find_optimal_discount(base_price, cogs, elasticity, optimize_for='revenue'):
    """
    หาส่วนลดที่ทำให้ revenue หรือ gross profit สูงสุด
    
    สมมติ baseline qty = 100 หน่วย (เราสนใจ % change เทียบ baseline)
    
    Revenue(d)     = price * (1-d) * qty(d)
    qty(d)         = qty_base * ((1-d))^elasticity   (จาก elasticity definition)
    GrossProfit(d) = (price*(1-d) - cogs) * qty(d)
    """
    discounts      = np.arange(0, 0.51, 0.01)  # 0% ถึง 50%
    base_qty       = 100  # baseline
    
    results = []
    for d in discounts:
        price_d  = base_price * (1 - d)
        # qty เพิ่มขึ้นเมื่อราคาลด (elasticity เป็นลบ)
        price_ratio = (1 - d)  # price_d / base_price
        qty_d    = base_qty * (price_ratio ** elasticity)
        revenue  = price_d * qty_d
        gp       = (price_d - cogs) * qty_d
        results.append({'discount': d, 'revenue': revenue, 'gross_profit': gp, 'qty': qty_d})
    
    res_df = pd.DataFrame(results)
    
    if optimize_for == 'revenue':
        best_idx = res_df['revenue'].idxmax()
    else:
        best_idx = res_df['gross_profit'].idxmax()
    
    return res_df, best_idx


# ========== 4.4 Compute Recommendations for All Products ==========

# Current promotion ของแต่ละสินค้า (ใช้โปรล่าสุดถ้ามี)
current_promo = (
    df[df['has_promotion'] == 1]
    .groupby('product_id')['discount']
    .mean()
    .reset_index()
    .rename(columns={'discount': 'current_discount'})
)

recommendations = []

for _, row in elasticity_df.iterrows():
    pid        = row['product_id']
    base_price = row['price']
    cogs       = row['cogs']
    elasticity = row['elasticity']
    
    # Skip ถ้าข้อมูลไม่ครบ
    if pd.isna(base_price) or pd.isna(cogs) or pd.isna(elasticity):
        continue
    
    # หา optimal discount (optimize Gross Profit)
    res_df, best_idx = find_optimal_discount(base_price, cogs, elasticity, optimize_for='gross_profit')
    best_row         = res_df.loc[best_idx]
    baseline_row     = res_df[res_df['discount'] == 0].iloc[0]
    
    # Current discount (ถ้าเคยมีโปร)
    cur_promo_row = current_promo[current_promo['product_id'] == pid]
    cur_discount  = cur_promo_row['current_discount'].values[0] if len(cur_promo_row) > 0 else 0.0
    
    # คำนวณ current revenue/gp
    cur_idx = res_df['discount'].sub(cur_discount).abs().idxmin()
    cur_row = res_df.loc[cur_idx]
    
    # Revenue/Profit change
    rev_change = (best_row['revenue'] - cur_row['revenue']) / (cur_row['revenue'] + 1e-9)
    gp_change  = (best_row['gross_profit'] - cur_row['gross_profit']) / (abs(cur_row['gross_profit']) + 1e-9)
    
    # Action
    if best_row['discount'] > cur_discount + 0.03:
        action = 'Increase Discount'
    elif best_row['discount'] < cur_discount - 0.03:
        action = 'Reduce Discount'
    else:
        action = 'Keep / Stop Promotion'
    
    recommendations.append({
        'product_id'           : pid,
        'category'             : row['product_taxonomies'],
        'list_price'           : round(base_price, 2),
        'cogs'                 : round(cogs, 2),
        'elasticity'           : round(elasticity, 3),
        'current_discount'     : round(cur_discount, 2),
        'recommended_discount' : round(best_row['discount'], 2),
        'expected_rev_change'  : f"{rev_change:+.1%}",
        'expected_gp_change'   : f"{gp_change:+.1%}",
        'action'               : action
    })

reco_df = pd.DataFrame(recommendations)
print(f'Generated recommendations for {len(reco_df)} products')
reco_df.head(10)
````

````python
# ========== Action Summary ==========

action_count = reco_df['action'].value_counts()
print('Action Summary:')
print(action_count)

fig, ax = plt.subplots(figsize=(7, 4))
colors_map = {
    'Increase Discount'   : '#4CAF50',
    'Reduce Discount'     : '#F44336',
    'Keep / Stop Promotion': '#9E9E9E'
}
bar_colors = [colors_map.get(a, 'steelblue') for a in action_count.index]
ax.bar(action_count.index, action_count.values, color=bar_colors, edgecolor='white')
ax.set_title('Recommended Actions across All Products')
ax.set_ylabel('Number of Products')
plt.tight_layout()
plt.show()
````

````python
# ========== 4.5 Visualize Optimal Discount Curve (ตัวอย่าง 2 สินค้า) ==========

# เลือกสินค้า elastic และ inelastic
elastic_prod   = reco_df[reco_df['elasticity'] < -1].iloc[0]
inelastic_prod = reco_df[reco_df['elasticity'].between(-1, 0)].iloc[0]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

for ax, prod, label in [
    (axes[0], elastic_prod,   'Elastic Product'),
    (axes[1], inelastic_prod, 'Inelastic Product')
]:
    res, best_idx = find_optimal_discount(
        prod['list_price'], prod['cogs'], prod['elasticity'], optimize_for='gross_profit'
    )
    ax.plot(res['discount']*100, res['gross_profit'], color='steelblue', label='Gross Profit')
    ax.plot(res['discount']*100, res['revenue'],      color='coral',     linestyle='--', label='Revenue')
    ax.axvline(res.loc[best_idx, 'discount']*100, color='green', linestyle=':', label=f'Optimal Discount')
    ax.set_title(f'{label}\n{prod["product_id"]} | Elasticity={prod["elasticity"]:.2f}')
    ax.set_xlabel('Discount (%)')
    ax.set_ylabel('Value (Index, baseline=100 units)')
    ax.legend(fontsize=8)

plt.suptitle('Gross Profit & Revenue vs Discount Level', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
````

## STEP 5: Evaluation

### วิธี Validate โดยไม่ต้องมี model สมบูรณ์

#### Method 1: Back-Testing บนข้อมูลเดิม
- แบ่ง transaction เป็น Train (80%) / Test (20%)
- Train elasticity บน 80% → Predict qty ใน 20%
- วัด MAPE (Mean Absolute % Error) ของ qty ที่ predict

#### Method 2: Simple A/B Comparison
- เลือกสินค้าที่โปรโมชั่นหลายครั้งในช่วง discount ต่างกัน
- เปรียบเทียบ Avg Revenue/Unit vs Avg Qty ในแต่ละช่วง
- ดูว่าตรงกับ Elasticity ที่เราคำนวณหรือเปล่า

#### Method 3: Sanity Check ทางธุรกิจ
- Promotion ROI > 1.0 = คุ้มทุน
- Gross Profit ไม่ต่ำกว่า Breakeven

````python
# ========== 5.1 Simple Back-test Validation ==========

# ใช้สินค้าที่มีทั้ง promo และ non-promo transactions
promo_products = df[df['has_promotion'] == 1]['product_id'].unique()

comparison_rows = []
for pid in promo_products:
    subset = df[df['product_id'] == pid]
    promo_subset   = subset[subset['has_promotion'] == 1]
    nopromo_subset = subset[subset['has_promotion'] == 0]
    
    if len(promo_subset) < 3 or len(nopromo_subset) < 3:
        continue
    
    avg_qty_promo    = promo_subset['qty'].mean()
    avg_qty_nopromo  = nopromo_subset['qty'].mean()
    avg_rev_promo    = promo_subset['revenue'].mean()
    avg_rev_nopromo  = nopromo_subset['revenue'].mean()
    avg_disc         = promo_subset['discount'].mean()
    
    qty_uplift = (avg_qty_promo - avg_qty_nopromo) / (avg_qty_nopromo + 1e-9)
    rev_uplift = (avg_rev_promo - avg_rev_nopromo) / (avg_rev_nopromo + 1e-9)
    
    comparison_rows.append({
        'product_id'     : pid,
        'avg_discount'   : round(avg_disc, 2),
        'qty_no_promo'   : round(avg_qty_nopromo, 1),
        'qty_with_promo' : round(avg_qty_promo, 1),
        'qty_uplift'     : f"{qty_uplift:+.1%}",
        'rev_uplift'     : f"{rev_uplift:+.1%}",
        'promo_beneficial': 'YES' if rev_uplift > 0 else 'NO'
    })

comparison_df = pd.DataFrame(comparison_rows)
print(f'Validated {len(comparison_df)} products with promo vs non-promo transactions')
print('\nSample:')
print(comparison_df.head(10).to_string())

beneficial_pct = (comparison_df['promo_beneficial'] == 'YES').mean()
print(f'\nPromotion was revenue-beneficial in {beneficial_pct:.0%} of tested products')
````

````python
# ========== 5.2 Promotion ROI Validation ==========

# Promotion ROI = Revenue Uplift / Discount Cost
# สินค้าที่มีข้อมูลทั้งคู่

roi_rows = []
for _, row in comparison_df.iterrows():
    pid      = row['product_id']
    disc     = row['avg_discount']
    
    subset         = df[df['product_id'] == pid]
    promo_subset   = subset[subset['has_promotion'] == 1]
    nopromo_subset = subset[subset['has_promotion'] == 0]
    
    avg_rev_promo   = promo_subset['revenue'].mean()
    avg_rev_nopromo = nopromo_subset['revenue'].mean()
    
    rev_uplift    = avg_rev_promo - avg_rev_nopromo
    discount_cost = avg_rev_promo * disc  # เงินที่เสียไปจากการลดราคา
    
    roi = rev_uplift / (discount_cost + 1e-9)
    roi_rows.append({'product_id': pid, 'avg_discount': disc, 'promo_roi': round(roi, 2)})

roi_df = pd.DataFrame(roi_rows)

fig, ax = plt.subplots(figsize=(8, 4))
roi_sorted = roi_df.sort_values('promo_roi')
colors = ['#d32f2f' if r < 1 else '#388e3c' for r in roi_sorted['promo_roi']]
ax.barh(roi_sorted['product_id'], roi_sorted['promo_roi'], color=colors)
ax.axvline(1, color='black', linestyle='--', label='Break-even ROI = 1.0')
ax.set_title('Promotion ROI by Product (Green = ROI > 1, \nRed = ROI < 1 = ไม่คุ้มลด)')
ax.set_xlabel('Promotion ROI')
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()

print(f"Avg Promo ROI: {roi_df['promo_roi'].mean():.2f}")
print(f"Products with ROI > 1: {(roi_df['promo_roi'] > 1).sum()} / {len(roi_df)}")
````

````python
# ========== 5.3 Baseline Comparison ==========
# เทียบ: ถ้าลูกค้าใช้ Recommended Discount vs Current Practice

# สินค้าที่แนะนำให้ Reduce Discount
reduce_prods = reco_df[reco_df['action'] == 'Reduce Discount']
increase_prods = reco_df[reco_df['action'] == 'Increase Discount']
keep_prods = reco_df[reco_df['action'] == 'Keep / Stop Promotion']

print('=== Recommendation Summary ===')
print(f"Reduce Discount   : {len(reduce_prods):>3} products")
print(f"Increase Discount : {len(increase_prods):>3} products")
print(f"Keep/Stop         : {len(keep_prods):>3} products")

print('\n=== Top 10 Products to Reduce Discount (Wasting Margin) ===')
print(
    reduce_prods[['product_id','category','elasticity','current_discount','recommended_discount','expected_gp_change']]
    .sort_values('current_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)

print('\n=== Top 10 Products to Increase Discount (Missed Opportunity) ===')
print(
    increase_prods[['product_id','category','elasticity','current_discount','recommended_discount','expected_rev_change']]
    .sort_values('recommended_discount', ascending=False)
    .head(10)
    .to_string(index=False)
)
````

## 📌 สรุป & ข้อจำกัด (Limitations & Assumptions)

### Assumptions
1. **Log-Log relationship** — สมมติว่า Demand function เป็น Power Law (เหมาะกับ Retail ทั่วไป)
2. **Stable Elasticity** — สมมติว่า Elasticity ไม่เปลี่ยนตามฤดูกาลหรือเศรษฐกิจ
3. **Ceteris Paribus** — สมมติว่าปัจจัยอื่นคงที่ (คู่แข่ง, สินค้าทดแทน)
4. **buy1get1 discount = 50%** — assumption เพื่อ normalize ค่า missing discount

### Limitations
- ข้อมูลมีเพียง 1 ปี → Elasticity อาจยังไม่เสถียร
- สินค้าที่มี transaction น้อยใช้ category median แทน → accuracy ต่ำกว่า
- ไม่ได้คำนึง cross-product effects (เช่น ลดสินค้า A อาจกระทบสินค้า B)
- store_master: ใช้เพื่อ segment analysis เท่านั้น ยังไม่ได้ optimize ระดับ store
