# 💡 Price Elasticity-Based Promotion Optimizer
**CRISP-DM Framework | Retail Revenue Maximization**

---

**Business Problem:**  
SME Retail กำลังถูกบีบจากต้นทุนสูงและสินค้าจีนราคาถูก  
การจัด Promotion แบบเดาสุ่มทำลาย Margin โดยไม่ช่วยเพิ่ม Revenue จริง  

**Solution:**  
ใช้ข้อมูลการขายจริงมาประมาณ Price Elasticity ของแต่ละสินค้า  
แล้วแนะนำส่วนลดที่ Maximize Revenue หรือ Gross Profit ได้จริง

**KPI หลัก:**  
- Gross Profit Margin  
- Promotion ROI  
- Revenue Uplift

## STEP 1: Business Understanding

### ปัญหาที่แท้จริงคืออะไร?

ร้าน Retail ขนาดเล็ก-กลาง (SME) มักจัดโปรโมชั่นแบบ **"จัดไปก่อน แล้วค่อยดูผล"**  
ซึ่งทำให้:
- บางสินค้าลด 20% แต่ยอดขายไม่ขึ้นเลย → Margin หาย เปล่าประโยชน์
- บางสินค้าลดแค่ 5% ก็ขายดีขึ้นมาก → แต่กลับไม่จัดโปรโมชั่น เสียโอกาส

### Solution ที่เสนอ: Price Elasticity-Based Optimizer

**ทำอะไร?**  
ประมาณค่า Price Elasticity of Demand (PED) ของแต่ละสินค้า จากข้อมูลขายจริง  
แล้วหาส่วนลดที่ทำให้ Revenue หรือ Gross Profit สูงสุด

```
Input:  Sales Transaction + Product Master + Promotion Master + Customer + Store
Output: Promotion Recommendation Table (ลดเท่าไหร่ → คาด Revenue/Profit เปลี่ยนแค่ไหน)
```

**ทำไม Rule-based ธรรมดาไม่พอ?**
- Rule-based: "ถ้าสต็อกเยอะ → ลด 20%" → ไม่ได้มองพฤติกรรมลูกค้าจริง
- Dashboard: บอกว่า "ยอดขายลด" แต่ไม่บอกว่า "ควรทำอะไร"
- Elasticity Model: **บอกได้ว่าลดเท่าไหร่ถึงคุ้ม** และผลต่อ Margin เป็นเท่าไหร่

### Stakeholders
- **Primary User:** เจ้าของร้าน / ผู้จัดการฝ่ายการตลาด
- **Decision:** จะจัดโปรโมชั่นสินค้าไหน ลดเท่าไหร่ ช่วงไหน

### KPI Definitions
| KPI | นิยาม | เป้าหมาย |
|-----|--------|----------|
| Gross Profit Margin | (Revenue - COGS) / Revenue | เพิ่มขึ้น vs baseline |
| Promotion ROI | Revenue Uplift / Discount Cost | > 1.0 (คุ้มทุน) |
| Revenue Uplift | % เปลี่ยนแปลงของยอดขาย vs ก่อนโปร | บวกและมีนัยสำคัญ |

## STEP 3: Data Cleaning & Preparation

#### 3.1 Clean sales_transaction

````python
print(f'Original Shape: {sales.shape}')

# 1. drop duplicate rows 
sales = sales.drop_duplicates().reset_index(drop=True)

# 2. parse datetime
sales['datetime'] = pd.to_datetime(sales['datetime'], errors='coerce')

# 3. keep only valid year 2024
sales = sales[sales['datetime'].dt.year == 2024].copy()

# 4. drop rows with missing price
sales = sales.dropna(subset=['price'])

# 5. remove negative quantity
sales = sales[sales['qty'] > 0].copy()

# 6. Remove quantity outliers
outlier_qty_values = [500, 888, 999]
sales = sales[~sales['qty'].isin(outlier_qty_values)].copy()

# 7. Create revenue column
sales['revenue'] = sales['price'] * sales['qty']

# 8. Create promotion flag
sales['has_promotion'] = sales['promotion_id'].notna().astype(int)

print(f'Cleaned Shape: {sales.shape}')
````

#### 3.2 Clean product_master

````python
print(f'Original Shape: {product.shape}')

# 1. drop duplicate rows 
product = product.drop_duplicates().reset_index(drop=True)

# 2. Fill null product_taxonomies to 'unknown'
product['product_taxonomies'] = product['product_taxonomies'].fillna('unknown')

# 3. Normalize product_taxonomies
product['product_taxonomies'] = (product['product_taxonomies'].str.strip().str.lower())

# 4. Strip 'baht' and whitespace from the price, then convert it to float
product['price'] = (
    product['price']
    .astype(str)
    .str.replace(r'[^\d.]', '', regex=True)
)

product['price'] = pd.to_numeric(product['price'], errors='coerce')

print(f'Cleaned Shape: {product.shape}')
````

#### 3.3 Clean promotion_master

````python
print(f'Original Shape: {promo.shape}')

promo['start_date'] = pd.to_datetime(
    promo['start_date'],
    errors='coerce'
)

promo['end_date'] = pd.to_datetime(
    promo['end_date'],
    errors='coerce'
)

print(f'Cleaned Shape: {promo.shape}')
````

````python
promo.dtypes
````

#### 3.4 Clean Store Master

````python
print(f'Original Shape: {store.shape}')

# Normalize store_taxonomies
store['store_taxonomies'] = (
    store['store_taxonomies']
    .str.strip()
    .str.lower()
    .str.replace(r'\s+store$', '', regex=True)
)

print(f'Cleaned Shape: {store.shape}')
````

````python
print('Unique Store Taxonomies:')
print(store['store_taxonomies'].value_counts())
````

#### 3.5 Clean Customer Master

````python
print(f'Original Shape: {customer.shape}')

# 1. Normalize customer_taxonomies
customer['customer_taxonomies'] = (customer['customer_taxonomies'].str.strip().str.lower())

# 2. Correct the spelling
typo_map = {'ocasional': 'occasional', 'loyl': 'loyal'}
customer['customer_taxonomies'] = customer['customer_taxonomies'].replace(typo_map)

# 3. Fill null customer_taxonomies with 'unknown'
customer['customer_taxonomies'] = customer['customer_taxonomies'].fillna('unknown')

# 4. Drop duplicate rows
customer = customer.drop_duplicates().reset_index(drop=True)

print(f'Cleaned Shape: {customer.shape}')
````

#### 3.6 Merge Table

````python
# Join sales & product (ได้ COGS, product category)
df = sales.merge(
    product[['product_id', 'price', 'product_taxonomies', 'cogs']],
    on='product_id',
    how='left',
    suffixes=('_sold', '_list') 
)

# Join with promotion (ได้ discount rate)
df = df.merge(
    promo[['promotion_id', 'discount', 'start_date', 'end_date']],
    on='promotion_id',
    how='left'
)

# Join with customer
df = df.merge(
    customer[['customer_id', 'customer_taxonomies']],
    on='customer_id',
    how='left'
)

# Join with store
df = df.merge(
    store[['store_id', 'store_taxonomies']],
    on='store_id',
    how='left'
)

print(df.shape)
print(df.head())
````

#### 3.7 Feature Engineering

````python
import pandas as pd
import numpy as np

# =========================================================
# STEP 3.7: Feature Engineering (เวอร์ชันแก้ไขสมบูรณ์)
# =========================================================

# 1. คำนวณฝั่งรายธุรกรรม (Transaction Level) ลงใน df
df['revenue'] = df['price_sold'] * df['qty']

# คำนวณส่วนลดจริง (และใช้ .clip เพื่อตัดกรณีขายแพงกว่าป้ายให้เป็น 0)
df['actual_discount_pct'] = np.where(
    df['price_list'] > 0, 
    (df['price_list'] - df['price_sold']) / df['price_list'], 
    0
)
df['actual_discount_pct'] = df['actual_discount_pct'].clip(lower=0)

# คำนวณกำไรขั้นต้นและ Margin
df['gross_profit'] = df['revenue'] - (df['cogs'] * df['qty'])
df['margin_pct'] = np.where(df['revenue'] > 0, df['gross_profit'] / df['revenue'], np.nan)

# 2. สร้าง Time Features สำหรับดักจับ Seasonality
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# 3. Aggregate เป็นรายสัปดาห์ (Product-Week Level) ด้วยวิธีการเฉลี่ยถ่วงน้ำหนัก
df['weighted_price_helper'] = df['price_sold'] * df['qty']
df['weighted_cogs_helper'] = df['cogs'] * df['qty']
df['promo_count_helper'] = df['has_promotion'] 

weekly_sales_v3 = df.set_index('datetime').groupby(
    ['product_id', pd.Grouper(freq='W')]
).agg({
    'qty': 'sum',
    'weighted_price_helper': 'sum',
    'weighted_cogs_helper': 'sum',
    'promo_count_helper': 'sum',      # จำนวนออเดอร์ที่ติดโปรในสัปดาห์นั้น
    'po_id': 'count'                  # จำนวนออเดอร์ทั้งหมดในสัปดาห์นั้น
}).reset_index()

# คำนวณราคากลับจากตัวช่วยถ่วงน้ำหนัก
weekly_sales_v3['price_sold'] = weekly_sales_v3['weighted_price_helper'] / weekly_sales_v3['qty']
weekly_sales_v3['cogs'] = weekly_sales_v3['weighted_cogs_helper'] / weekly_sales_v3['qty']

# คำนวณอัตราความเข้มข้นของโปรโมชั่น (Promo Rate)
weekly_sales_v3['promo_rate'] = weekly_sales_v3['promo_count_helper'] / weekly_sales_v3['po_id']

# ลบคอลัมน์ Helper ที่ไม่ใช้แล้วออก
weekly_sales_v3 = weekly_sales_v3.drop(columns=['weighted_price_helper', 'weighted_cogs_helper', 'promo_count_helper', 'po_id'])

# กรองเฉพาะสัปดาห์ที่มีการขายจริง
weekly_sales_v3 = weekly_sales_v3[(weekly_sales_v3['qty'] > 0) & (weekly_sales_v3['price_sold'] > 0)]


# =========================================================
# STEP 3.8: โค้ดเช็คสถิติ (Sanity Check V3)
# =========================================================

print("--- 1. ข้อมูลเชิงสถิติของ Margin และ Discount ---")
print(df[['actual_discount_pct', 'gross_profit', 'margin_pct']].describe().round(2))

negative_margin = df[df['margin_pct'] < 0]
print(f"\nจำนวนรายการที่ขายขาดทุน (Margin ติดลบ): {len(negative_margin)} รายการ (คิดเป็น {len(negative_margin)/len(df):.1%})")

if len(negative_margin) > 0:
    print("\nตัวอย่างรายการที่ขายขาดทุน:")
    print(negative_margin[['product_id', 'price_sold', 'cogs', 'qty', 'actual_discount_pct', 'margin_pct']].head())

print("\n--- 2. สัดส่วนการขายแบ่งตามวันในสัปดาห์ (0=จันทร์, 6=อาทิตย์) ---")
print(df['day_of_week'].value_counts().sort_index())
print(f"สัดส่วนยอดขายวันหยุด (Weekend): {df['is_weekend'].mean():.1%}")

print(f"\n--- 3. ขนาดของตาราง weekly_sales_v3: {weekly_sales_v3.shape} ---")
print("ตัวอย่างข้อมูล 5 แถวแรก:")
print(weekly_sales_v3[['product_id', 'datetime', 'qty', 'price_sold', 'cogs', 'promo_rate']].head())

print("\n--- สถิติของความเข้มข้นโปรโมชั่น (Promo Rate) ---")
print(weekly_sales_v3['promo_rate'].describe().round(2))

weeks_per_product = weekly_sales_v3.groupby('product_id').size()
print("\n--- สถิติจำนวนสัปดาห์ที่มีการขาย ต่อ 1 สินค้า ---")
print(weeks_per_product.describe().round(1))

ready_products = (weeks_per_product > 10).sum()
print(f"\nจำนวนสินค้าที่มีข้อมูลพร้อมรันโมเดล (>10 สัปดาห์): {ready_products} จาก {len(weeks_per_product)} รายการ")
````

## STEP 4 : Model Selection & Experimenting

````python
# =========================================================
# STEP 4 : Model Selection & Experimenting (อัปเดตตามคำแนะนำ)
# =========================================================
import statsmodels.api as sm

def categorize_elasticity(coef):
    if coef < -1:
        return 'elastic'
    else:
        return 'inelastic'

# 1. จัดเตรียม Log Data รายสัปดาห์
weekly_sales_v3['log_qty'] = np.log(weekly_sales_v3['qty'])
weekly_sales_v3['log_price'] = np.log(weekly_sales_v3['price_sold'])

prod_cat_map = df.groupby('product_id')['product_taxonomies'].first().to_dict()

# --- หาค่า Elasticity เฉลี่ยราย Category ไว้เป็น Fallback ---
temp_results = []
for pid in weekly_sales_v3['product_id'].unique():
    product_data = weekly_sales_v3[weekly_sales_v3['product_id'] == pid]
    X = product_data[['log_price', 'promo_rate']]
    X = sm.add_constant(X)
    y = product_data['log_qty']
    try:
        model = sm.OLS(y, X).fit()
        # เก็บเฉพาะตัวที่ P-value น่าเชื่อถือและลดราคาแล้วยอดขายขึ้น (e < 0)
        if model.pvalues['log_price'] <= 0.05 and model.params['log_price'] < 0:
            temp_results.append({'product_id': pid, 'ped': model.params['log_price']})
    except:
        continue

# สร้าง Dictionary สำหรับ Fallback
if len(temp_results) > 0:
    temp_df = pd.DataFrame(temp_results).merge(df[['product_id', 'product_taxonomies']].drop_duplicates(), on='product_id')
    category_fallback = temp_df.groupby('product_taxonomies')['ped'].mean().to_dict()
else:
    category_fallback = {}

# 2. วนลูปสกัด Elasticity ทุกสินค้ารายตัว
elasticity_results_v3 = []
for pid in weekly_sales_v3['product_id'].unique():
    product_data = weekly_sales_v3[weekly_sales_v3['product_id'] == pid]
    cat = prod_cat_map.get(pid, 'unknown')
    
    X = product_data[['log_price', 'promo_rate']]
    X = sm.add_constant(X)
    y = product_data['log_qty']
    
    try:
        model = sm.OLS(y, X).fit()
        ped = model.params['log_price']
        p_value = model.pvalues['log_price']
        r_squared = model.rsquared
        
        # --- Statistical Filtering & Fallback ---
        # ถ้าค่า P-value ขาดความน่าเชื่อถือ หรือค่า PED เป็นบวก (ลดราคาแล้วขายแย่ลง) ให้ใช้ค่าเฉลี่ย Category
        if p_value > 0.05 or ped >= 0:
            ped = category_fallback.get(cat, -0.5) # ถ้าหมวดหมู่นั้นไม่มีข้อมูลเลย ให้ใช้ -0.5 ไว้ก่อน
            
        elasticity_results_v3.append({
            'product_id': pid,
            'elasticity_coef': ped,
            'p_value': p_value, # บันทึก p_value เดิมไว้ดู
            'r_squared': r_squared,
            'avg_price': product_data['price_sold'].mean(),
            'avg_cogs': product_data['cogs'].mean(),
            'avg_weekly_qty': product_data['qty'].mean()
        })
    except:
        continue

df_ped_v3 = pd.DataFrame(elasticity_results_v3)

# --- Winsorization: ตัด Outlier ที่แรงเกินไป ---
# บีบกรอบให้อยู่ระหว่าง -3.0 ถึง -0.1 ตามบริบท Retail FMCG ปกติ
df_ped_v3['elasticity_coef'] = df_ped_v3['elasticity_coef'].clip(-3.0, -0.1)

df_ped_v3['category_elasticity'] = df_ped_v3['elasticity_coef'].apply(categorize_elasticity)
````

## STEP 5 : Evaluation

````python
# =========================================================
# STEP 5 : Evaluation (เวอร์ชันแก้ไขสมบูรณ์ - ปรับสูตรและขยาย Clipping)
# =========================================================
sim_results = []

for _, row in df_ped_v3.iterrows():
    pid = row['product_id']
    e = row['elasticity_coef']
    p_val = row['p_value']
    
    # คำนวณอัตรากำไรพื้นฐาน (Margin Rate)
    margin_rate = (row['avg_price'] - row['avg_cogs']) / row['avg_price']
    
    # ดึงค่าส่วนลดปัจจุบัน (Current Discount) เฉลี่ยจริงจากประวัติขาย
    current_discount = df[df['product_id'] == pid]['actual_discount_pct'].mean()
    
    # --- 1. คำนวณราคาส่วนลดแนะนำทางทฤษฎี (Theoretical Rec Discount) ---
    if p_val <= 0.05 and e < -1:
        # 🌟 [จุดแก้ไขที่ 1]: เปลี่ยนมาใช้สูตร Log-Log Derivation ที่ถูกต้องตามหลักเศรษฐศาสตร์
        rec_discount = (1 + (e * margin_rate)) / (1 + e)
        
        # คุมกรอบส่วนลดแนะนำไม่ให้เทกระจายเกิน 40% เพื่อความปลอดภัยของธุรกิจ
        rec_discount = np.clip(rec_discount, 0.0, 0.40)
        elasticity_label = "🔵 elastic"
    else:
        rec_discount = 0.0
        elasticity_label = "🔴 inelastic"
        # บังคับค่า e สำหรับจำลองฝั่งไร้ความไว ไม่ให้ยอดขายหดตัวเกินจริง
        e = max(e, -0.5) if e < 0 else -0.1

    # --- 2. ใส่ Business Rule Layer เพื่อกำหนด Action (รวม Keep Current) ---
    discount_diff = rec_discount - current_discount
    
    if elasticity_label == "🔵 elastic":
        if abs(discount_diff) <= 0.05:
            action = "🟢 Keep Current"
            rec_discount = current_discount 
        elif discount_diff > 0.05:
            action = "🔺 Increase Discount"
        else:
            action = "🔻 Reduce Discount"
            
    else: # ฝั่ง inelastic (rec_discount = 0.0)
        if current_discount <= 0.02: 
            action = "🟢 Keep Current"
            rec_discount = current_discount 
        else:
            action = "🔻 Reduce Discount"
            rec_discount = 0.0

    # --- 3. คำนวณสูตรพยากรณ์ผลกระทบ % Delta ด้วย Log-Log Formula ---
    # 💡 ปรับปรุง Logic: ถ้าเลือก Keep Current ผลกระทบต้องเป็น 0 เสมอตามจริง
    if action == "🟢 Keep Current":
        price_change_pct = 0.0
        qty_change_pct = 0.0
        rev_change_pct = 0.0
        profit_change_pct = 0.0
    else:
        price_ratio = (1 - rec_discount) / (1 - current_discount) if (1 - current_discount) > 0 else 1
        
        if price_ratio <= 0:
            price_ratio = 0.01 

        # การเปลี่ยนแปลงราคาและจำนวนชิ้น
        price_change_pct = price_ratio - 1
        qty_change_pct = (price_ratio ** e) - 1
        
        # คำนวณรายได้ใหม่
        rev_change_pct = (1 + price_change_pct) * (1 + qty_change_pct) - 1
        
        # 💡 ปรับปรุงสูตร Margin ใหม่: คำนวณจาก Price Ratio เทียบกับ Margin ปัจจุบันที่ถูกต้อง
        old_profit_margin = margin_rate
        new_profit_margin = 1 - ((1 - margin_rate) / price_ratio)
        
        if old_profit_margin > 0:
            profit_change_pct = ((1 + qty_change_pct) * price_ratio * new_profit_margin / old_profit_margin) - 1
        else:
            profit_change_pct = 0.0

    # ดึงหมวดหมู่สินค้า
    cat_name = prod_cat_map.get(pid, 'Unknown')
    
    sim_results.append({
        'Product ID': pid,
        'Category': cat_name,
        'Cur %': current_discount,
        'Rec %': rec_discount,
        'Rev Δ%': rev_change_pct,
        'Profit Δ%': profit_change_pct,
        'Action': action,
        'Elasticity': elasticity_label
    })

# แปลงเป็น DataFrame และแต่งขอบเขตความเหวี่ยงข้อมูลให้สมจริงในเชิงธุรกิจ (Business Cap)
df_simulation = pd.DataFrame(sim_results)
df_simulation['Profit Δ%'] = df_simulation['Profit Δ%'].clip(lower=-0.95, upper=1.00)
df_simulation['Rev Δ%'] = df_simulation['Rev Δ%'].clip(lower=-0.95, upper=1.00)
````

## STEP 6: Deployment

````python
# =========================================================
# STEP 6: Deployment & Export Recommendation Table
# =========================================================

# 1. ปริ้นต์หน้าจอรายงานสรุปฉบับผู้บริหาร (Top 15 Sorted by Profit Growth)
df_top15_recommendation = df_simulation.sort_values(by='Profit Δ%', ascending=False).head(15)

print("=" * 105)
print("PROMOTION RECOMMENDATION TABLE  (Top 15 by Expected Profit Change)")
print("=" * 105)
print(f"{'Product ID':<12}{'Category':<18}{'Cur %':<8}{'Rec %':<8}{'Rev Δ%':<10}{'Profit Δ%':<11}{'Action':<26}{'Elasticity':<12}")
print("-" * 105)

for _, row in df_top15_recommendation.iterrows():
    rev_sign = '+' if row['Rev Δ%'] >= 0 else ''
    prof_sign = '+' if row['Profit Δ%'] >= 0 else ''
    
    print(f"{row['Product ID']:<12}"
          f"{row['Category']:<18}"
          f"{row['Cur %']*100:>5.1f}%   "
          f"{row['Rec %']*100:>5.0f}%   "
          f"{rev_sign}{row['Rev Δ%']*100:>5.1f}%    "
          f"{prof_sign}{row['Profit Δ%']*100:>5.1f}%    "
          f"{row['Action']:<26}"
          f"{row['Elasticity']:<12}")
print("=" * 105)
````

## Visualize Baseline vs Model (Gross Profit)

````python
# =========================================================
# STEP 7: Visualize Baseline vs Model (Gross Profit)
# =========================================================
import matplotlib.pyplot as plt
import os

# --- 1. แปลง % Profit ให้เป็นตัวเงิน (THB) ประเมินจากยอดขายทั้งปี ---
# สมมติว่ายอดขายเฉลี่ยรายสัปดาห์ (avg_weekly_qty) คงที่ตลอด 52 สัปดาห์
df_simulation['Base_Profit_THB'] = (df_ped_v3['avg_price'] - df_ped_v3['avg_cogs']) * df_ped_v3['avg_weekly_qty'] * 52

# คำนวณกำไรเมื่อใช้โมเดลของเรา (Elasticity-Optimized)
df_simulation['Model_Profit_THB'] = df_simulation['Base_Profit_THB'] * (1 + df_simulation['Profit Δ%'])
total_model = df_simulation['Model_Profit_THB'].sum()

# --- 2. คำนวณกำไรของ Baseline (สมมติจัดโปรลด 15% เท่ากันทั้งร้าน) ---
total_baseline = 0
for _, row in df_ped_v3.iterrows():
    pid = row['product_id']
    e = max(row['elasticity_coef'], -0.5) if row['elasticity_coef'] > -1 else row['elasticity_coef']
    margin_rate = (row['avg_price'] - row['avg_cogs']) / row['avg_price']
    
    flat_discount = 0.15 # Baseline ลด 15%
    current_discount = df[df['product_id'] == pid]['actual_discount_pct'].mean()
    
    # คำนวณผลกระทบถ้าลด 15% ทุกตัว
    price_change_pct = ((1 - flat_discount) / (1 - current_discount)) - 1 if (1 - current_discount) > 0 else 0
    qty_change_pct = price_change_pct * e
    new_margin = 1 - ((1 - margin_rate) / (1 - flat_discount)) if (1 - flat_discount) > 0 else 0
    
    profit_change_pct = ((1 + qty_change_pct) * new_margin / margin_rate) - 1 if margin_rate > 0 else 0
    
    base_profit = (row['avg_price'] - row['avg_cogs']) * row['avg_weekly_qty'] * 52
    total_baseline += base_profit * (1 + profit_change_pct)

# --- 3. คำนวณ % การเติบโตรวม ---
improvement = ((total_model - total_baseline) / total_baseline) * 100

# --- 4. วาดกราฟ (ใช้โค้ดของคุณ) ---
labels   = ['Baseline\n(Flat 15%)', 'Model\n(Elasticity-Optimized)']
values   = [total_baseline, total_model]
bar_cols = ['#E8E8E8', '#4C8BF5']

plt.figure(figsize=(7, 5))
bars = plt.bar(labels, values, color=bar_cols, edgecolor='white', width=0.45)
plt.ylabel('Estimated Total Gross Profit (THB)')
plt.title('Baseline vs Optimized Promotion Strategy\n(Simulated on Mock Data)', fontsize=13)

for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() * 1.01,
             f'{val:,.0f}', ha='center', fontsize=11, fontweight='bold')

diff_label = f'+{improvement:.1f}%' if improvement >= 0 else f'{improvement:.1f}%'
plt.annotate(f'Improvement: {diff_label}',
             xy=(1, total_model),
             xytext=(0.5, max(values) * 0.55),
             fontsize=12, color='#2C5F9E', fontweight='bold',
             arrowprops=dict(arrowstyle='->', color='#2C5F9E'))

# ตกแต่งกราฟให้คลีนขึ้น
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
````

````python
# =========================================================
# STEP 8: Elasticity Curve (Visualizing the Optimum)
# =========================================================
def plot_elasticity_curve(pid):
    # 1. ดึงข้อมูลจากโมเดล
    product_stats = df_ped_v3[df_ped_v3['product_id'] == pid].iloc[0]
    e = product_stats['elasticity_coef']
    base_margin = (product_stats['avg_price'] - product_stats['avg_cogs']) / product_stats['avg_price']
    
    # 2. จำลองส่วนลดตั้งแต่ 0% ถึง 50%
    discounts = np.linspace(0, 0.50, 50)
    profit_curve = []
    
    for d in discounts:
        # คำนวณกำไรตามสูตร Elasticity
        # กฎ: กำไร = (P*(1-d) - COGS) * Q * (1 - d*e)
        # ปรับสูตรให้สอดคล้องกับพฤติกรรมลูกค้าจริง
        price_effect = (1 - d)
        qty_effect = (1 + (d * abs(e))) # ถ้า e เป็นลบ ให้ใช้ค่าสัมบูรณ์
        
        # คำนวณ Margin ต่อหน่วย
        margin_per_unit = (product_stats['avg_price'] * (1 - d)) - product_stats['avg_cogs']
        total_profit = margin_per_unit * (product_stats['avg_weekly_qty'] * qty_effect)
        profit_curve.append(total_profit)
        
    # 3. หาจุด Optimum
    opt_idx = np.argmax(profit_curve)
    opt_discount = discounts[opt_idx]
    opt_profit = profit_curve[opt_idx]
    
    # 4. วาดกราฟ
    plt.figure(figsize=(8, 5))
    plt.plot(discounts * 100, profit_curve, color='#4C8BF5', linewidth=3, label='Profit Curve')
    plt.scatter(opt_discount * 100, opt_profit, color='red', s=100, zorder=5, label=f'Optimum: {opt_discount*100:.1f}%')
    
    plt.title(f'Profit Optimization Curve: Product {pid}\n(Elasticity: {e:.2f})', fontsize=14)
    plt.xlabel('Discount Percentage (%)')
    plt.ylabel('Estimated Weekly Profit (THB)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    
    # ใส่ลูกศรชี้จุด Optimum
    plt.annotate(f'Max Profit\n{opt_discount*100:.0f}% Discount', 
                 xy=(opt_discount*100, opt_profit), 
                 xytext=(opt_discount*100 + 5, opt_profit * 0.9),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    
    plt.tight_layout()
    plt.savefig(f'./output/fig_07_elasticity_curve_{pid}.png', dpi=120)
    plt.show()

# รันกราฟสำหรับ P004
plot_elasticity_curve('P004')
````
